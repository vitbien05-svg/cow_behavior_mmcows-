```python
import os
import json
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score
# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CONFIG_PATH = r"d:\archive\mmcows\configs\config_s1.json"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')
# KhoбәЈng thб»қi gian tб»« бәЈnh
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"
# Tham sб»‘ mГҙ hГ¬nh & huбәҘn luyб»Үn
WINDOW_SIZE = 32       # Дҗб»ҷ dГ i chuб»—i (sб»‘ bЖ°б»ӣc 15s liГӘn tiбәҝp Д‘Ж°a vГ o BiLSTM)
PROJECTION_DIM = 256   # Sб»‘ chiб»Ғu Д‘бә§u ra cб»§a nhГЎnh sensor (yГӘu cбә§u cб»§a bбәЎn)
N_CLASSES = 7          # 7 hГ nh vi chГӯnh (bб»Ҹ behavior 0)
EPOCHS = 15
BATCH_SIZE = 32
def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())
START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)
# ================= 2. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE CHO Tб»ӘNG CON BГ’ =================
def load_and_align_cow_data(cow_id):
    """Дҗб»Қc vГ  gб»ҷp tбәҘt cбәЈ sensor cho 1 con bГІ, Д‘б»“ng bб»ҷ vб»Ғ lЖ°б»ӣi 15s"""
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        # Load tб»« cache nбәҝu Д‘ГЈ xб»ӯ lГҪ trЖ°б»ӣc Д‘Гі
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined
    # Дҗб»Қc 6 sensor vГ  nhГЈn behavior
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    # NhГЈn behavior (lбәҘy nhГЈn 1 Hz)
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)
    # LЖ°б»ӣi thб»қi gian 15s chuбә©n tб»« Start -> End
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    # HГ m merge
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df
    # Дҗб»“ng bб»ҷ hГіa
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    # Дҗiб»Ғn giГЎ trб»Ӣ trб»‘ng
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    # LЖ°u cache
    master_df.to_csv(cache_path, index=False)
    return master_df
# Pre-load/align cho cбәЈ 10 con bГІ
for cid in range(1, 11):
    load_and_align_cow_data(cid)
# ================= 3. HГҖM TДӮNG CЖҜб»ңNG Dб»® LIб»ҶU (DATA AUGMENTATION) =================
def augment_data(X):
    """
    ГҒp dб»Ҙng 2 kб»№ thuбәӯt Augmentation phб»• biбәҝn cho dб»Ҝ liб»Үu sensor:
    1. Jittering: ThГӘm nhiб»…u Gaussian ngбә«u nhiГӘn
    2. Scaling: NhГўn tб»· lб»Ү ngбә«u nhiГӘn cho biГӘn Д‘б»ҷ tГӯn hiб»Үu
    """
    # 1. Jittering
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    
    # 2. Scaling
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    
    # TrбәЈ vб»Ғ gб»ҷp cбәЈ dб»Ҝ liб»Үu gб»‘c vГ  dб»Ҝ liб»Үu Д‘ГЈ Д‘Ж°б»Јc tДғng cЖ°б»қng
    return np.concatenate([X, X_jitter, X_scaled], axis=0)
# ================= 4. PHГӮN ДҗOбә N CHUб»–I (SEGMENTATION & WINDOW SHIFT) =================
def segment_cow_data(df, window_size, overlap=0.0):
    """CбәҜt dб»Ҝ liб»Үu chuб»—i dГ i cб»§a bГІ thГ nh cГЎc Д‘oбәЎn window_size bЖ°б»ӣc"""
    # TГЎch Д‘бә·c trЖ°ng vГ  nhГЈn
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list = []
    y_list = []
    
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        # LoбәЎi bб»Ҹ cГЎc chuб»—i chб»©a nhГЈn 0 (unknown) Д‘б»ғ khб»ӣp vб»ӣi baseline cб»§a bбәЎn
        # (hoбә·c bбәЎn cГі thб»ғ giб»Ҝ tГ№y thuб»ҷc cГЎch Д‘б»Ӣnh nghД©a nhГЈn cб»§a dataset)
        if 0 not in y_seg:
            X_list.append(X_seg)
            y_list.append(y_seg - 1)  # GiбәЈm 1 Д‘ЖЎn vб»Ӣ Д‘б»ғ nhГЈn tб»« 1->7 chuyб»ғn vб»Ғ 0->6
            
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size))
        
    return np.array(X_list), np.array(y_list)
# ================= 5. KHб»һI Tбә O MГ” HГҢNH SENSOR BILSTM Vб»ҡI ДҗбәҰU RA 256 =================
def build_bilstm_model(seq_len, n_features, n_classes, projection_dim=256):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    
    # 1. NhГЎnh trГӯch xuбәҘt Д‘бә·c trЖ°ng BiLSTM cб»§a Sensor
    lstm_out = Bidirectional(LSTM(64, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    # 2. Lб»ӣp FC2 ГЎnh xбәЎ lГӘn 256 chiб»Ғu (yГӘu cбә§u Д‘бә§u ra lГ  256 cб»§a bбәЎn)
    fc2_out = Dense(projection_dim, activation='relu', name="FC2_Projection")(lstm_out)
    fc2_out = BatchNormalization()(fc2_out)
    fc2_out = Dropout(0.3)(fc2_out)
    
    # Lб»ӣp Д‘бә§u ra Д‘б»ғ train & Д‘ГЎnh giГЎ phГўn loбәЎi (Sequence-to-Sequence)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(fc2_out)
    
    # TбәЎo model hoГ n chб»үnh Д‘б»ғ huбәҘn luyб»Үn Д‘б»ҷc lбәӯp
    model = Model(inputs=inputs, outputs=outputs, name="Sensor_BiLSTM_Classifier")
    
    # Дҗб»“ng thб»қi trГӯch xuбәҘt riГӘng Feature Extractor (Д‘бә§u ra shape 256) Д‘б»ғ bбәЎn tГӯch hб»Јp sau nГ y
    feature_extractor = Model(inputs=inputs, outputs=fc2_out, name="Sensor_Feature_Extractor_256")
    
    return model, feature_extractor
# ================= 6. VГ’NG HUбәӨN LUYб»ҶN 5-FOLD CROSS-VALIDATION (6-2-2 SPLIT) =================
# Load danh sГЎch phГўn chia fold tб»« file json cб»§a dб»ұ ГЎn
with open(CONFIG_PATH, "r") as f:
    config_s1 = json.load(f)
folds = config_s1["folds"]
aggregated_f1 = []
for fold_name, fold_config in folds.items():
    print(f"\n==================== HUбәӨN LUYб»ҶN {fold_name.upper()} ====================")
    train_ids = [int(x) for x in fold_config["train"]]
    val_ids = [int(x) for x in fold_config["val"]]
    test_ids = [int(x) for x in fold_config["test"]]
    
    print(f"Train Cows: {train_ids} (6 con)")
    print(f"Val Cows: {val_ids} (2 con)")
    print(f"Test Cows: {test_ids} (2 con)")
    
    # Load vГ  gб»ҷp dб»Ҝ liб»Үu
    def get_data_for_cows(ids, overlap=0.0):
        X_all, y_all = [], []
        for cid in ids:
            df = load_and_align_cow_data(cid)
            X_cow, y_cow = segment_cow_data(df, WINDOW_SIZE, overlap)
            if len(X_cow) > 0:
                X_all.append(X_cow)
                y_all.append(y_cow)
        return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0)
    # Train: sб»ӯ dб»Ҙng overlap 50% Д‘б»ғ sinh nhiб»Ғu mбә«u hЖЎn (TДғng cЖ°б»қng dб»Ҝ liб»Үu thб»қi gian)
    X_train, y_train = get_data_for_cows(train_ids, overlap=0.5)
    # Val/Test: sб»ӯ dб»Ҙng overlap 0% Д‘б»ғ Д‘ГЎnh giГЎ cГҙng bбәұng
    X_val, y_val = get_data_for_cows(val_ids, overlap=0.0)
    X_test, y_test = get_data_for_cows(test_ids, overlap=0.0)
    
    # ГҒp dб»Ҙng Augmentation cho tбәӯp Train
    if len(X_train) > 0:
        X_train_augmented = augment_data(X_train)
        y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0) # duplicate nhГЈn tЖ°ЖЎng б»©ng
        print(f"Sб»‘ lЖ°б»Јng mбә«u Train sau khi Augment: {X_train.shape[0]} -> {X_train_augmented.shape[0]}")
        X_train, y_train = X_train_augmented, y_train_augmented
    # Chuбә©n hГіa (Fit scaler trГӘn tбәӯp Train, Transform trГӘn Val/Test)
    num_samples, num_timesteps, num_features = X_train.shape
    scaler = StandardScaler()
    X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
    X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
    
    if len(X_val) > 0:
        X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], WINDOW_SIZE, num_features)
    if len(X_test) > 0:
        X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], WINDOW_SIZE, num_features)
        
    # Chuyб»ғn Д‘б»•i nhГЈn sang One-hot
    y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
    y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
    y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
    
    # Khб»ҹi tбәЎo mГҙ hГ¬nh
    model, extractor = build_bilstm_model(WINDOW_SIZE, num_features, N_CLASSES, projection_dim=PROJECTION_DIM)
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics=["accuracy"]
    )
    
    # Callbacks
    checkpoint_path = f"best_model_{fold_name}.h5"
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
        ModelCheckpoint(checkpoint_path, monitor="val_loss", save_best_only=True)
    ]
    
    # Fit model
    history = model.fit(
        X_train, y_train_cat,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_val, y_val_cat),
        callbacks=callbacks,
        verbose=1
    )
    
    # ДҗГЎnh giГЎ trГӘn tбәӯp test Д‘б»ҷc lбәӯp cб»§a fold
    y_pred_cat = model.predict(X_test)
    y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
    y_true = np.argmax(y_test_cat, axis=-1).flatten()
    
    fold_f1 = f1_score(y_true, y_pred, average="macro")
    fold_acc = accuracy_score(y_true, y_pred)
    aggregated_f1.append(fold_f1)
    
    print(f"\nKбәҫT QUбәў {fold_name.upper()}:")
    print(f"- Accuracy: {fold_acc:.4f}")
    print(f"- Macro F1-Score: {fold_f1:.4f}")
    
    # Kiб»ғm tra kГӯch thЖ°б»ӣc Д‘бә§u ra cб»§a Feature Extractor (nhГЎnh 256 chiб»Ғu)
    feat_out = extractor.predict(X_test[:2])
    print(f"- Kiб»ғm tra shape feature extractor Д‘бә§u ra 256: {feat_out.shape} -> (Batch, Window, 256) thГ nh cГҙng!")
print("\n==================== KбәҫT QUбәў CHUNG 5-FOLD ====================")
print(f"Macro F1 trung bГ¬nh: {np.mean(aggregated_f1):.4f} Вұ {np.std(aggregated_f1):.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

# KhoбәЈng thб»қi gian tб»« бәЈnh
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

# Tham sб»‘ mГҙ hГ¬nh & huбәҘn luyб»Үn
WINDOW_SIZE = 32       # Дҗб»ҷ dГ i chuб»—i (sб»‘ bЖ°б»ӣc 15s liГӘn tiбәҝp Д‘Ж°a vГ o BiLSTM)
PROJECTION_DIM = 256   # Sб»‘ chiб»Ғu Д‘бә§u ra cб»§a nhГЎnh sensor (yГӘu cбә§u cб»§a bбәЎn)
N_CLASSES = 7          # 7 hГ nh vi chГӯnh (bб»Ҹ behavior 0)
EPOCHS = 15
BATCH_SIZE = 32

# ================= 2. Дҗб»ҠNH NGHДЁA 5-FOLD (CHIA THEO CON BГ’ 6 - 2 - 2) =================
FOLDS = {
    "fold_1": {
        "train": [1, 2, 3, 4, 5, 6],
        "val": [7, 8],
        "test": [9, 10]
    },
    "fold_2": {
        "train": [3, 4, 5, 6, 7, 8],
        "val": [9, 10],
        "test": [1, 2]
    },
    "fold_3": {
        "train": [5, 6, 7, 8, 9, 10],
        "val": [1, 2],
        "test": [3, 4]
    },
    "fold_4": {
        "train": [7, 8, 9, 10, 1, 2],
        "val": [3, 4],
        "test": [5, 6]
    },
    "fold_5": {
        "train": [9, 10, 1, 2, 3, 4],
        "val": [5, 6],
        "test": [7, 8]
    }
}

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE CHO Tб»ӘNG CON BГ’ =================
def load_and_align_cow_data(cow_id):
    """Дҗб»Қc vГ  gб»ҷp tбәҘt cбәЈ sensor cho 1 con bГІ, Д‘б»“ng bб»ҷ vб»Ғ lЖ°б»ӣi 15s"""
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    # Дҗб»Қc 6 sensor vГ  nhГЈn behavior
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    # LЖ°б»ӣi thб»қi gian 15s chuбә©n tб»« Start -> End
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    # HГ m merge
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    # Дҗб»“ng bб»ҷ hГіa
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    # Дҗiб»Ғn giГЎ trб»Ӣ trб»‘ng
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# TбәЎo sбәөn cache cho cбәЈ 10 con bГІ
print("--- KHб»һI Tбә O BЖҜб»ҡC Дҗб»’NG Bб»ҳ VГҖ CACHE Dб»® LIб»ҶU ---")
for cid in range(1, 11):
    load_and_align_cow_data(cid)

# ================= 4. PHГӮN ДҗOбә N CHUб»–I (SEGMENTATION & WINDOW SHIFT - KHГ”NG OVERLAP) =================
def segment_cow_data(df, window_size, overlap=0.0):
    """CбәҜt dб»Ҝ liб»Үu chuб»—i dГ i cб»§a bГІ thГ nh cГЎc Д‘oбәЎn window_size bЖ°б»ӣc"""
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list = []
    y_list = []
    w_list = [] # sample weights Д‘б»ғ bб»Ҹ qua nhГЈn 0 khi tГӯnh Loss
    
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0) # dummy class
                w_seg.append(0.0)  # ignore loss (trб»Қng sб»‘ 0)
            else:
                y_mapped.append(val - 1) # chuyб»ғn Д‘б»•i nhГЈn 1..7 vб»Ғ 0..6
                w_seg.append(1.0)        # tГӯnh loss bГ¬nh thЖ°б»қng
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
        
    return np.array(X_list), np.array(y_list), np.array(w_list)

# ================= 5. KHб»һI Tбә O MГ” HГҢNH SENSOR BILSTM Vб»ҡI ДҗбәҰU RA 256 =================
def build_bilstm_model(seq_len, n_features, n_classes, projection_dim=256):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    
    # 1. NhГЎnh trГӯch xuбәҘt Д‘бә·c trЖ°ng BiLSTM cб»§a Sensor
    lstm_out = Bidirectional(LSTM(64, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    # 2. Lб»ӣp FC2 ГЎnh xбәЎ lГӘn 256 chiб»Ғu
    fc2_out = Dense(projection_dim, activation='relu', name="FC2_Projection")(lstm_out)
    fc2_out = BatchNormalization()(fc2_out)
    fc2_out = Dropout(0.3)(fc2_out)
    
    # Lб»ӣp phГўn loбәЎi Sequence-to-Sequence (DГ№ng Д‘б»ғ train Д‘б»ҷc lбәӯp nhГЎnh sensor)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(fc2_out)
    
    # TбәЎo model hoГ n chб»үnh Д‘б»ғ huбәҘn luyб»Үn Д‘б»ҷc lбәӯp
    model = Model(inputs=inputs, outputs=outputs, name="Sensor_BiLSTM_Classifier")
    
    # Дҗб»“ng thб»қi trГӯch xuбәҘt riГӘng Feature Extractor (Д‘бә§u ra shape 256)
    feature_extractor = Model(inputs=inputs, outputs=fc2_out, name="Sensor_Feature_Extractor_256")
    
    return model, feature_extractor

# ================= 6. VГ’NG HUбәӨN LUYб»ҶN 5-FOLD CROSS-VALIDATION (KHГ”NG AUGMENTATION) =================
aggregated_f1 = []

for fold_name, fold_config in FOLDS.items():
    print(f"\n==================== HUбәӨN LUYб»ҶN {fold_name.upper()} (NO AUGMENTATION) ====================")
    train_ids = fold_config["train"]
    val_ids = fold_config["val"]
    test_ids = fold_config["test"]
    
    print(f"Train Cows: {train_ids} (6 con)")
    print(f"Val Cows: {val_ids} (2 con)")
    print(f"Test Cows: {test_ids} (2 con)")
    
    # Load vГ  phГўn Д‘oбәЎn dб»Ҝ liб»Үu
    def get_data_for_cows(ids, overlap=0.0):
        X_all, y_all, w_all = [], [], []
        for cid in ids:
            df = load_and_align_cow_data(cid)
            X_cow, y_cow, w_cow = segment_cow_data(df, WINDOW_SIZE, overlap)
            if len(X_cow) > 0:
                X_all.append(X_cow)
                y_all.append(y_cow)
                w_all.append(w_cow)
        return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

    # Train/Val/Test: Дҗб»Ғu dГ№ng overlap = 0.0 (KhГҙng sinh thГӘm dб»Ҝ liб»Үu vГ  khГҙng augment)
    X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=0.0)
    X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0)
    X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0)
    
    print(f"Sб»‘ lЖ°б»Јng mбә«u Train thб»ұc tбәҝ (KhГҙng overlap, khГҙng noise): {X_train.shape[0]}")

    # Chuбә©n hГіa (Fit scaler trГӘn tбәӯp Train, Transform trГӘn Val/Test)
    num_samples, num_timesteps, num_features = X_train.shape
    scaler = StandardScaler()
    X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
    X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
    
    if len(X_val) > 0:
        X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], WINDOW_SIZE, num_features)
    if len(X_test) > 0:
        X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], WINDOW_SIZE, num_features)
        
    # Chuyб»ғn Д‘б»•i nhГЈn sang One-hot
    y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
    y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
    y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
    
    # Khб»ҹi tбәЎo mГҙ hГ¬nh
    model, extractor = build_bilstm_model(WINDOW_SIZE, num_features, N_CLASSES, projection_dim=PROJECTION_DIM)
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics=["accuracy"]
    )
    
    # Callbacks
    checkpoint_path = f"best_model_no_aug_{fold_name}.h5"
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
        ModelCheckpoint(checkpoint_path, monitor="val_loss", save_best_only=True)
    ]
    
    # Fit model (sample_weight loбәЎi bб»Ҹ nhГЈn 0 khi tГӯnh Loss)
    history = model.fit(
        X_train, y_train_cat,
        sample_weight=w_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_val, y_val_cat, w_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # ДҗГЎnh giГЎ trГӘn tбәӯp test
    y_pred_cat = model.predict(X_test)
    y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
    y_true = np.argmax(y_test_cat, axis=-1).flatten()
    w_flat = w_test.flatten()
    
    # Chб»ү Д‘ГЎnh giГЎ trГӘn cГЎc bЖ°б»ӣc thб»қi gian cГі nhГЈn hб»Јp lб»Ү (w == 1.0)
    y_pred_valid = y_pred[w_flat == 1.0]
    y_true_valid = y_true[w_flat == 1.0]
    
    fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
    fold_acc = accuracy_score(y_true_valid, y_pred_valid)
    aggregated_f1.append(fold_f1)
    
    print(f"\nKбәҫT QUбәў Dб»° ДҗOГҒN {fold_name.upper()} (NO AUGMENTATION):")
    print(f"- Accuracy: {fold_acc:.4f}")
    print(f"- Macro F1-Score: {fold_f1:.4f}")

print("\n==================== KбәҫT QUбәў CHUNG 5-FOLD (NO AUGMENTATION) ====================")
print(f"Macro F1 trung bГ¬nh: {np.mean(aggregated_f1):.4f} Вұ {np.std(aggregated_f1):.4f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout

def create_sensor_bilstm_branch(window_size, n_features, projection_dim=256):
    """
    TбәЎo nhГЎnh BiLSTM trГӯch xuбәҘt Д‘бә·c trЖ°ng Sensor (UWB, IMMU, ANKLE, ANKLE_ACCEL, HEAD_DIRECT, PRESSURE)
    - window_size: Sб»‘ lЖ°б»Јng bЖ°б»ӣc thб»қi gian 15s (Sequence Length - tЖ°ЖЎng б»©ng X_0, X_1... X_t)
    - n_features: Tб»•ng sб»‘ Д‘бә·c trЖ°ng cб»§a 6 sensor (21 Д‘бә·c trЖ°ng)
    - projection_dim: Sб»‘ chiб»Ғu Д‘бә§u ra sau lб»ӣp FC2 (yГӘu cбә§u lГ  256 chiб»Ғu)
    """
    # 1. Дҗб»Ӣnh nghД©a Д‘бә§u vГ o: Shape (Batch, Window, Features)
    inputs = Input(shape=(window_size, n_features), name="sensor_input")
    
    # 2. Lб»ӣp BiLSTM: return_sequences=True Д‘б»ғ giб»Ҝ lбәЎi chuб»—i thб»қi gian khб»ӣp vб»ӣi Camera
    # Дҗбә§u ra cб»§a BiLSTM: (Batch, Window, Hidden_Units * 2) = (Batch, Window, 128)
    lstm_out = Bidirectional(LSTM(64, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    # 3. Lб»ӣp Fully Connected 2 (FC2) chiбәҝu sб»‘ chiб»Ғu vб»Ғ 256
    # Дҗбә§u ra: (Batch, Window, 256)
    fc2_out = Dense(projection_dim, activation='relu', name="FC2_Projection")(lstm_out)
    fc2_out = BatchNormalization(name="FC2_BN")(fc2_out)
    fc2_out = Dropout(0.3, name="FC2_Dropout")(fc2_out)
    
    # Khб»ҹi tбәЎo Keras Model
    model = Model(inputs=inputs, outputs=fc2_out, name="Sensor_BiLSTM_Branch")
    
    return model

# ================= CHбә Y THб»¬ Vб»ҡI Dб»® LIб»ҶU GIбәў Lбә¬P =================
if __name__ == "__main__":
    # 1. Thiбәҝt lбәӯp tham sб»‘ mбә«u
    BATCH_SIZE = 8       # Sб»‘ lЖ°б»Јng chuб»—i xб»ӯ lГҪ song song
    WINDOW_SIZE = 5039   # Sб»‘ bЖ°б»ӣc thб»қi gian 15s tЖ°ЖЎng б»©ng khoбәЈng thб»қi gian trong бәЈnh (21 tiбәҝng)
    N_FEATURES = 21      # Tб»•ng 21 Д‘бә·c trЖ°ng cб»§a 6 cбәЈm biбәҝn ghГ©p lбәЎi
    PROJECTION_DIM = 256 # Sб»‘ chiб»Ғu Д‘бә§u ra yГӘu cбә§u
    
    # 2. Khб»ҹi tбәЎo mГҙ hГ¬nh
    model = create_sensor_bilstm_branch(window_size=WINDOW_SIZE, n_features=N_FEATURES, projection_dim=PROJECTION_DIM)
    
    # In bбәЈng kiбәҝn trГәc tГіm tбәҜt cб»§a mГҙ hГ¬nh
    model.summary()
    
    # 3. TбәЎo dб»Ҝ liб»Үu giбәЈ lбәӯp Д‘бә§u vГ o (Mock Input)
    # Shape: (8, 5039, 21)
    mock_sensor_input = np.random.rand(BATCH_SIZE, WINDOW_SIZE, N_FEATURES)
    print(f"\n[INPUT] HГ¬nh dГЎng dб»Ҝ liб»Үu sensor Д‘бә§u vГ o: {mock_sensor_input.shape}")
    
    # 4. ChбәЎy mГҙ hГ¬nh Д‘б»ғ lбәҘy Дҗбә·c trЖ°ng Д‘бә§u ra (Forward Pass)
    features_out = model.predict(mock_sensor_input)
    
    # 5. In kбәҝt quбәЈ hГ¬nh dГЎng Д‘бә§u ra
    print(f"[OUTPUT] HГ¬nh dГЎng Д‘бә·c trЖ°ng Д‘бә§u ra cб»§a nhГЎnh sensor: {features_out.shape}")
    print(f"-> GiбәЈi thГӯch: (Batch_Size={features_out.shape[0]}, Window_Size={features_out.shape[1]}, Features={features_out.shape[2]})")
```

```python
import os
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

# KhoбәЈng thб»қi gian tб»« бәЈnh
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

# Tham sб»‘ mГҙ hГ¬nh & huбәҘn luyб»Үn
WINDOW_SIZE = 32       # Дҗб»ҷ dГ i chuб»—i (sб»‘ bЖ°б»ӣc 15s liГӘn tiбәҝp Д‘Ж°a vГ o BiLSTM)
N_CLASSES = 7          # 7 hГ nh vi chГӯnh (bб»Ҹ behavior 0)
EPOCHS = 15
BATCH_SIZE = 32

# ================= 2. Дҗб»ҠNH NGHДЁA 5-FOLD (CHIA THEO CON BГ’ 6 - 2 - 2) =================
# CГі 10 con bГІ (ID tб»« 1 Д‘бәҝn 10). Mб»—i fold chб»Қn:
# - 6 con lГ m Train
# - 2 con lГ m Val
# - 2 con lГ m Test
FOLDS = {
    "fold_1": {
        "train": [1, 2, 3, 4, 5, 6],
        "val": [7, 8],
        "test": [9, 10]
    },
    "fold_2": {
        "train": [3, 4, 5, 6, 7, 8],
        "val": [9, 10],
        "test": [1, 2]
    },
    "fold_3": {
        "train": [5, 6, 7, 8, 9, 10],
        "val": [1, 2],
        "test": [3, 4]
    },
    "fold_4": {
        "train": [7, 8, 9, 10, 1, 2],
        "val": [3, 4],
        "test": [5, 6]
    },
    "fold_5": {
        "train": [9, 10, 1, 2, 3, 4],
        "val": [5, 6],
        "test": [7, 8]
    }
}

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE CHO Tб»ӘNG CON BГ’ =================
def load_and_align_cow_data(cow_id):
    """Дҗб»Қc vГ  gб»ҷp tбәҘt cбәЈ sensor cho 1 con bГІ, Д‘б»“ng bб»ҷ vб»Ғ lЖ°б»ӣi 15s"""
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    # Дҗб»Қc 6 sensor vГ  nhГЈn behavior
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    # LЖ°б»ӣi thб»қi gian 15s chuбә©n tб»« Start -> End
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    # HГ m merge
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    # Дҗб»“ng bб»ҷ hГіa
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    # Дҗiб»Ғn giГЎ trб»Ӣ trб»‘ng
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# TбәЎo sбәөn cache cho cбәЈ 10 con bГІ
print("--- KHб»һI Tбә O BЖҜб»ҡC Дҗб»’NG Bб»ҳ VГҖ CACHE Dб»® LIб»ҶU ---")
for cid in range(1, 11):
    load_and_align_cow_data(cid)

# ================= 4. HГҖM TДӮNG CЖҜб»ңNG Dб»® LIб»ҶU (DATA AUGMENTATION) =================
def augment_data(X):
    """
    ГҒp dб»Ҙng 2 kб»№ thuбәӯt Augmentation phб»• biбәҝn cho dб»Ҝ liб»Үu sensor:
    1. Jittering: ThГӘm nhiб»…u Gaussian ngбә«u nhiГӘn
    2. Scaling: NhГўn tб»· lб»Ү ngбә«u nhiГӘn cho biГӘn Д‘б»ҷ tГӯn hiб»Үu
    """
    # Jittering
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    
    # Scaling
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    
    # TrбәЈ vб»Ғ gб»ҷp cбәЈ dб»Ҝ liб»Үu gб»‘c vГ  dб»Ҝ liб»Үu Д‘ГЈ tДғng cЖ°б»қng
    return np.concatenate([X, X_jitter, X_scaled], axis=0)

# ================= 5. PHГӮN ДҗOбә N CHUб»–I (SEGMENTATION & WINDOW SHIFT) =================
def segment_cow_data(df, window_size, overlap=0.0):
    """CбәҜt dб»Ҝ liб»Үu chuб»—i dГ i cб»§a bГІ thГ nh cГЎc Д‘oбәЎn window_size bЖ°б»ӣc"""
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list = []
    y_list = []
    w_list = [] # sample weights Д‘б»ғ bб»Ҹ qua nhГЈn 0 khi tГӯnh Loss
    
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0) # dummy class
                w_seg.append(0.0)  # ignore loss (trб»Қng sб»‘ 0)
            else:
                y_mapped.append(val - 1) # chuyб»ғn Д‘б»•i nhГЈn 1..7 vб»Ғ 0..6
                w_seg.append(1.0)        # tГӯnh loss bГ¬nh thЖ°б»қng
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
        
    return np.array(X_list), np.array(y_list), np.array(w_list)

# ================= 6. KHб»һI Tбә O MГ” HГҢNH SENSOR BILSTM ДҗбәҰU RA 256 ДҗГғ LЖҜб»ўC Bб»Һ FC2 =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    
    # 1. NhГЎnh BiLSTM cб»§a Sensor: 128 units -> Дҗбә§u ra trб»ұc tiбәҝp lГ  256 chiб»Ғu (128 * 2)
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    # ГҒp dб»Ҙng trб»ұc tiбәҝp Regularization lГӘn Д‘бә§u ra BiLSTM Д‘б»ғ train б»•n Д‘б»Ӣnh
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    
    # Lб»ӣp phГўn loбәЎi Sequence-to-Sequence (DГ№ng Д‘б»ғ train Д‘б»ҷc lбәӯp nhГЎnh sensor)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    
    # TбәЎo model hoГ n chб»үnh Д‘б»ғ huбәҘn luyб»Үn Д‘б»ҷc lбәӯp
    model = Model(inputs=inputs, outputs=outputs, name="Sensor_BiLSTM_Classifier")
    
    # Дҗб»“ng thб»қi trГӯch xuбәҘt riГӘng Feature Extractor (Д‘бә§u ra shape 256 cб»§a lб»ӣp BiLSTM_Core)
    feature_extractor = Model(inputs=inputs, outputs=lstm_out, name="Sensor_Feature_Extractor_256")
    
    return model, feature_extractor

# ================= 7. VГ’NG HUбәӨN LUYб»ҶN 5-FOLD CROSS-VALIDATION =================
aggregated_f1 = []

for fold_name, fold_config in FOLDS.items():
    print(f"\n==================== HUбәӨN LUYб»ҶN {fold_name.upper()} ====================")
    train_ids = fold_config["train"]
    val_ids = fold_config["val"]
    test_ids = fold_config["test"]
    
    print(f"Train Cows: {train_ids} (6 con)")
    print(f"Val Cows: {val_ids} (2 con)")
    print(f"Test Cows: {test_ids} (2 con)")
    
    # Load vГ  phГўn Д‘oбәЎn dб»Ҝ liб»Үu
    def get_data_for_cows(ids, overlap=0.0):
        X_all, y_all, w_all = [], [], []
        for cid in ids:
            df = load_and_align_cow_data(cid)
            X_cow, y_cow, w_cow = segment_cow_data(df, WINDOW_SIZE, overlap)
            if len(X_cow) > 0:
                X_all.append(X_cow)
                y_all.append(y_cow)
                w_all.append(w_cow)
        return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

    # Train: sб»ӯ dб»Ҙng overlap 50% Д‘б»ғ sinh nhiб»Ғu mбә«u hЖЎn (TДғng cЖ°б»қng dб»Ҝ liб»Үu thб»қi gian)
    X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=0.5)
    # Val/Test: sб»ӯ dб»Ҙng overlap 0% Д‘б»ғ Д‘ГЎnh giГЎ cГҙng bбәұng
    X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0)
    X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0)
    
    # ГҒp dб»Ҙng Augmentation cho tбәӯp Train
    if len(X_train) > 0:
        X_train_augmented = augment_data(X_train)
        y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0)
        w_train_augmented = np.concatenate([w_train, w_train, w_train], axis=0)
        print(f"Sб»‘ lЖ°б»Јng mбә«u Train sau khi Augment: {X_train.shape[0]} -> {X_train_augmented.shape[0]}")
        X_train, y_train, w_train = X_train_augmented, y_train_augmented, w_train_augmented

    # Chuбә©n hГіa (Fit scaler trГӘn tбәӯp Train, Transform trГӘn Val/Test)
    num_samples, num_timesteps, num_features = X_train.shape
    scaler = StandardScaler()
    X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
    X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
    
    if len(X_val) > 0:
        X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], WINDOW_SIZE, num_features)
    if len(X_test) > 0:
        X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], WINDOW_SIZE, num_features)
        
    # Chuyб»ғn Д‘б»•i nhГЈn sang One-hot
    y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
    y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
    y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
    
    # Khб»ҹi tбәЎo mГҙ hГ¬nh
    model, extractor = build_bilstm_model(WINDOW_SIZE, num_features, N_CLASSES)
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics=["accuracy"]
    )
    
    # Callbacks
    checkpoint_path = f"best_model_{fold_name}.h5"
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
        ModelCheckpoint(checkpoint_path, monitor="val_loss", save_best_only=True)
    ]
    
    # Fit model (truyб»Ғn sample_weight Д‘б»ғ bб»Ҹ qua nhГЈn 0 khi tГӯnh Loss)
    history = model.fit(
        X_train, y_train_cat,
        sample_weight=w_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_val, y_val_cat, w_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # ДҗГЎnh giГЎ trГӘn tбәӯp test Д‘б»ҷc lбәӯp cб»§a fold
    y_pred_cat = model.predict(X_test)
    y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
    y_true = np.argmax(y_test_cat, axis=-1).flatten()
    w_flat = w_test.flatten()
    
    # Chб»ү Д‘ГЎnh giГЎ trГӘn cГЎc bЖ°б»ӣc thб»қi gian cГі nhГЈn hб»Јp lб»Ү (w == 1.0)
    y_pred_valid = y_pred[w_flat == 1.0]
    y_true_valid = y_true[w_flat == 1.0]
    
    fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
    fold_acc = accuracy_score(y_true_valid, y_pred_valid)
    aggregated_f1.append(fold_f1)
    
    print(f"\nKбәҫT QUбәў Dб»° ДҗOГҒN {fold_name.upper()}:")
    print(f"- Accuracy (chб»ү tГӯnh nhГЈn hб»Јp lб»Ү): {fold_acc:.4f}")
    print(f"- Macro F1-Score (chб»ү tГӯnh nhГЈn hб»Јp lб»Ү): {fold_f1:.4f}")
    
    # Kiб»ғm tra kГӯch thЖ°б»ӣc Д‘бә§u ra cб»§a Feature Extractor (nhГЎnh BiLSTM trб»ұc tiбәҝp)
    feat_out = extractor.predict(X_test[:2])
    print(f"- Kiб»ғm tra shape feature extractor Д‘бә§u ra 256 tб»« BiLSTM trб»ұc tiбәҝp: {feat_out.shape} -> (Batch, Window, 256) thГ nh cГҙng!")

print("\n==================== KбәҫT QUбәў CHUNG 5-FOLD ====================")
print(f"Macro F1 trung bГ¬nh: {np.mean(aggregated_f1):.4f} Вұ {np.std(aggregated_f1):.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

# KhoбәЈng thб»қi gian tб»« бәЈnh
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

# Tham sб»‘ mГҙ hГ¬nh & huбәҘn luyб»Үn
WINDOW_SIZE = 32       # Дҗб»ҷ dГ i chuб»—i (sб»‘ bЖ°б»ӣc 15s liГӘn tiбәҝp Д‘Ж°a vГ o BiLSTM)
N_CLASSES = 7          # 7 hГ nh vi chГӯnh (bб»Ҹ behavior 0)
EPOCHS = 10            # Дҗб»ғ Epoch = 10 Д‘б»ғ chбәЎy so sГЎnh nhanh hЖЎn
BATCH_SIZE = 32

# ================= 2. Дҗб»ҠNH NGHДЁA 5-FOLD (CHIA THEO CON BГ’ 6 - 2 - 2) =================
FOLDS = {
    "fold_1": {"train": [1, 2, 3, 4, 5, 6], "val": [7, 8], "test": [9, 10]},
    "fold_2": {"train": [3, 4, 5, 6, 7, 8], "val": [9, 10], "test": [1, 2]},
    "fold_3": {"train": [5, 6, 7, 8, 9, 10], "val": [1, 2], "test": [3, 4]},
    "fold_4": {"train": [7, 8, 9, 10, 1, 2], "val": [3, 4], "test": [5, 6]},
    "fold_5": {"train": [9, 10, 1, 2, 3, 4], "val": [5, 6], "test": [7, 8]}
}

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE CHO Tб»ӘNG CON BГ’ =================
def load_and_align_cow_data(cow_id):
    """Дҗб»Қc vГ  gб»ҷp tбәҘt cбәЈ sensor cho 1 con bГІ, Д‘б»“ng bб»ҷ vб»Ғ lЖ°б»ӣi 15s"""
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    # Дҗб»Қc 6 sensor vГ  nhГЈn behavior
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    # LЖ°б»ӣi thб»қi gian 15s chuбә©n tб»« Start -> End
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    # HГ m merge
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    # Дҗб»“ng bб»ҷ hГіa
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    # Дҗiб»Ғn giГЎ trб»Ӣ trб»‘ng
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# ================= 4. HГҖM TДӮNG CЖҜб»ңNG Dб»® LIб»ҶU (DATA AUGMENTATION) =================
def augment_data(X):
    # Jittering
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    # Scaling
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    return np.concatenate([X, X_jitter, X_scaled], axis=0)

# ================= 5. PHГӮN ДҗOбә N CHUб»–I (SEGMENTATION) =================
def segment_cow_data(df, window_size, overlap=0.0):
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list, y_list, w_list = [], [], []
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0)
                w_seg.append(0.0)
            else:
                y_mapped.append(val - 1)
                w_seg.append(1.0)
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
    return np.array(X_list), np.array(y_list), np.array(w_list)

# ================= 6. KHб»һI Tбә O MГ” HГҢNH SENSOR BILSTM Vб»ҡI ДҗбәҰU RA DIRECT 256 =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    # BiLSTM sб»ӯ dб»Ҙng 128 units -> Дҗбә§u ra 2 chiб»Ғu ghГ©p lбәЎi lГ  256 chiб»Ғu
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    model = Model(inputs=inputs, outputs=outputs, name="Sensor_BiLSTM_Classifier")
    return model

# ================= 7. VГ’NG HUбәӨN LUYб»ҶN CHUNG CHO Cбәў 2 PHЖҜЖ NG ГҒN =================
def run_evaluation(use_augmentation=True):
    print(f"\n=======================================================")
    print(f"Bбә®T ДҗбәҰU CHбә Y HUбәӨN LUYб»ҶN: {'CГ“ AUGMENTATION' if use_augmentation else 'KHГ”NG CГ“ AUGMENTATION'}")
    print(f"=======================================================")
    
    fold_f1_scores = []
    
    for fold_name, fold_config in FOLDS.items():
        train_ids = fold_config["train"]
        val_ids = fold_config["val"]
        test_ids = fold_config["test"]
        
        def get_data_for_cows(ids, overlap=0.0):
            X_all, y_all, w_all = [], [], []
            for cid in ids:
                df = load_and_align_cow_data(cid)
                X_cow, y_cow, w_cow = segment_cow_data(df, WINDOW_SIZE, overlap)
                if len(X_cow) > 0:
                    X_all.append(X_cow)
                    y_all.append(y_cow)
                    w_all.append(w_cow)
            return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

        # CбәҘu hГ¬nh Overlap phб»Ҙ thuб»ҷc vГ o Augmentation
        train_overlap = 0.5 if use_augmentation else 0.0
        
        X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=train_overlap)
        X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0)
        X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0)
        
        # ГҒp dб»Ҙng Jittering & Scaling nбәҝu chб»Қn Augmentation
        if use_augmentation and len(X_train) > 0:
            X_train_augmented = augment_data(X_train)
            y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0)
            w_train_augmented = np.concatenate([w_train, w_train, w_train], axis=0)
            X_train, y_train, w_train = X_train_augmented, y_train_augmented, w_train_augmented

        # Chuбә©n hГіa
        num_samples, num_timesteps, num_features = X_train.shape
        scaler = StandardScaler()
        X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
        X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
        
        if len(X_val) > 0:
            X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], WINDOW_SIZE, num_features)
        if len(X_test) > 0:
            X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], WINDOW_SIZE, num_features)
            
        y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
        y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
        y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
        
        model = build_bilstm_model(WINDOW_SIZE, num_features, N_CLASSES)
        model.compile(
            loss="categorical_crossentropy",
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
            metrics=["accuracy"]
        )
        
        checkpoint_name = f"best_model_{'aug' if use_augmentation else 'no_aug'}_{fold_name}.h5"
        callbacks = [
            EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True),
            ModelCheckpoint(checkpoint_name, monitor="val_loss", save_best_only=True)
        ]
        
        # Fit model
        model.fit(
            X_train, y_train_cat,
            sample_weight=w_train,
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_val, y_val_cat, w_val),
            callbacks=callbacks,
            verbose=0  # бәЁn bб»ӣt log train tб»«ng epoch cho Д‘б»Ў rб»‘i mбәҜt
        )
        
        # ДҗГЎnh giГЎ trГӘn tбәӯp test
        y_pred_cat = model.predict(X_test, verbose=0)
        y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
        y_true = np.argmax(y_test_cat, axis=-1).flatten()
        w_flat = w_test.flatten()
        
        y_pred_valid = y_pred[w_flat == 1.0]
        y_true_valid = y_true[w_flat == 1.0]
        
        fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
        fold_f1_scores.append(fold_f1)
        print(f"  - {fold_name.upper()}: Test Macro F1 = {fold_f1:.4f}")
        
    mean_f1 = np.mean(fold_f1_scores)
    std_f1 = np.std(fold_f1_scores)
    return mean_f1, std_f1

# ================= 8. CHЖҜЖ NG TRГҢNH CHГҚNH =================
if __name__ == "__main__":
    print("--- KHб»һI Tбә O BЖҜб»ҡC Дҗб»’NG Bб»ҳ VГҖ CACHE Dб»® LIб»ҶU ---")
    for cid in range(1, 11):
        load_and_align_cow_data(cid)
        
    # 1. ChбәЎy phЖ°ЖЎng ГЎn KHГ”NG cГі Augmentation (Baseline)
    f1_no_aug, std_no_aug = run_evaluation(use_augmentation=False)
    
    # 2. ChбәЎy phЖ°ЖЎng ГЎn CГ“ Augmentation
    f1_with_aug, std_with_aug = run_evaluation(use_augmentation=True)
    
    # 3. In kбәҝt quбәЈ Д‘б»‘i chiбәҝu cuб»‘i cГ№ng
    print("\n=======================================================")
    print("================ KбәҫT QUбәў Дҗб»җI CHIбәҫU CUб»җI CГҷNG ================")
    print("=======================================================")
    print(f"- KHГ”NG CГ“ Augmentation:  Macro F1 = {f1_no_aug*100:.2f}% Вұ {std_no_aug*100:.2f}%")
    print(f"- CГ“ CГ“ DГҷNG Augmentation: Macro F1 = {f1_with_aug*100:.2f}% Вұ {std_with_aug*100:.2f}%")
    
    diff = (f1_with_aug - f1_no_aug) * 100
    print(f"- Sб»ұ chГӘnh lб»Үch hiб»Үu nДғng: {diff:+.2f}%")
    print("=======================================================")
```

```python
import os
import json
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score
import matplotlib.pyplot as plt

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

# KhoбәЈng thб»қi gian tб»« бәЈnh
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

# Hyperparameter Search Space
WINDOW_SIZES = [16, 32, 64] # CГЎc kГӯch thЖ°б»ӣc cб»ӯa sб»• Д‘б»ғ tГ¬m tб»‘i Ж°u
MAX_EPOCHS = 100            # ChбәЎy tб»‘i Д‘a 100 epochs
BATCH_SIZE = 32
N_CLASSES = 7

# ================= 2. Дҗб»ҠNH NGHДЁA FOLDS (6 - 2 - 2 SPLIT) =================
FOLDS = {
    "fold_1": {"train": [1, 2, 3, 4, 5, 6], "val": [7, 8], "test": [9, 10]},
    "fold_2": {"train": [3, 4, 5, 6, 7, 8], "val": [9, 10], "test": [1, 2]},
    "fold_3": {"train": [5, 6, 7, 8, 9, 10], "val": [1, 2], "test": [3, 4]},
    "fold_4": {"train": [7, 8, 9, 10, 1, 2], "val": [3, 4], "test": [5, 6]},
    "fold_5": {"train": [9, 10, 1, 2, 3, 4], "val": [5, 6], "test": [7, 8]}
}

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE =================
def load_and_align_cow_data(cow_id):
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# ================= 4. DATA AUGMENTATION =================
def augment_data(X):
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    return np.concatenate([X, X_jitter, X_scaled], axis=0)

# ================= 5. SEGMENTATION =================
def segment_cow_data(df, window_size, overlap=0.0):
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list, y_list, w_list = [], [], []
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0)
                w_seg.append(0.0)
            else:
                y_mapped.append(val - 1)
                w_seg.append(1.0)
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
    return np.array(X_list), np.array(y_list), np.array(w_list)

# ================= 6. BILSTM MODEL =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    # BiLSTM sб»ӯ dб»Ҙng 128 units -> Дҗбә§u ra 256 chiб»Ғu trб»ұc tiбәҝp trЖ°б»ӣc phГўn loбәЎi
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# ================= 7. HUбәӨN LUYб»ҶN VГҖ TГҢM KIбәҫM WINDOW SIZE Tб»җI ЖҜU =================
results_by_window = {}
histories_by_window = {}

if __name__ == "__main__":
    print("--- KHб»һI Tбә O CACHE Dб»® LIб»ҶU ---")
    for cid in range(1, 11):
        load_and_align_cow_data(cid)
        
    for w_size in WINDOW_SIZES:
        print(f"\n=======================================================")
        print(f"THб»¬ NGHIб»ҶM WINDOW SIZE = {w_size} (Mб»‘c 15s)")
        print(f"=======================================================")
        
        fold_f1_scores = []
        fold_histories = []
        
        for fold_name, fold_config in FOLDS.items():
            print(f"\n--- {fold_name.upper()} (w_size={w_size}) ---")
            train_ids = fold_config["train"]
            val_ids = fold_config["val"]
            test_ids = fold_config["test"]
            
            def get_data_for_cows(ids, overlap=0.0):
                X_all, y_all, w_all = [], [], []
                for cid in ids:
                    df = load_and_align_cow_data(cid)
                    X_cow, y_cow, w_cow = segment_cow_data(df, w_size, overlap)
                    if len(X_cow) > 0:
                        X_all.append(X_cow)
                        y_all.append(y_cow)
                        w_all.append(w_cow)
                if len(X_all) == 0:
                    return np.empty((0, w_size, 21)), np.empty((0, w_size)), np.empty((0, w_size))
                return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

            # CбәҜt chuб»—i & ГҒp dб»Ҙng Augmentation cho Train
            X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=0.5)
            X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0)
            X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0)
            
            if len(X_train) > 0:
                X_train_augmented = augment_data(X_train)
                y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0)
                w_train_augmented = np.concatenate([w_train, w_train, w_train], axis=0)
                X_train, y_train, w_train = X_train_augmented, y_train_augmented, w_train_augmented

            # Chuбә©n hГіa
            num_samples, num_timesteps, num_features = X_train.shape
            scaler = StandardScaler()
            X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
            X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
            
            if len(X_val) > 0:
                X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], w_size, num_features)
            if len(X_test) > 0:
                X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], w_size, num_features)
                
            y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
            y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
            y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
            
            model = build_bilstm_model(w_size, num_features, N_CLASSES)
            model.compile(
                loss="categorical_crossentropy",
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                metrics=["accuracy"]
            )
            
            # Early stopping kiб»ғm soГЎt tб»‘i Д‘a 100 Epochs. Patience = 8.
            early_stopping = EarlyStopping(
                monitor="val_loss", 
                patience=8, 
                restore_best_weights=True,
                verbose=1
            )
            
            # Fit model
            history = model.fit(
                X_train, y_train_cat,
                sample_weight=w_train,
                epochs=MAX_EPOCHS,
                batch_size=BATCH_SIZE,
                validation_data=(X_val, y_val_cat, w_val),
                callbacks=[early_stopping],
                verbose=1
            )
            
            # ДҗГЎnh giГЎ
            y_pred_cat = model.predict(X_test)
            y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
            y_true = np.argmax(y_test_cat, axis=-1).flatten()
            w_flat = w_test.flatten()
            
            y_pred_valid = y_pred[w_flat == 1.0]
            y_true_valid = y_true[w_flat == 1.0]
            
            fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
            fold_f1_scores.append(fold_f1)
            
            fold_histories.append({
                'loss': history.history['loss'],
                'val_loss': history.history['val_loss']
            })
            
        # LЖ°u kбәҝt quбәЈ trung bГ¬nh cб»§a window_size nГ y
        mean_f1 = np.mean(fold_f1_scores)
        results_by_window[w_size] = mean_f1
        histories_by_window[w_size] = fold_histories
        print(f"\n=> Kбәҝt quбәЈ trung bГ¬nh cho Window Size {w_size}: F1-Score = {mean_f1*100:.2f}%")

    # ================= 8. TГҢM RA THAM Sб»җ Tб»җI ЖҜU NHбәӨT VГҖ Vбәј BIб»ӮU Дҗб»’ =================
    best_w_size = max(results_by_window, key=results_by_window.get)
    print(f"\n=======================================================")
    print(f"KбәҫT QUбәў Tб»җI ЖҜU NHбәӨT:")
    print(f"- Window Size tб»‘t nhбәҘt: {best_w_size}")
    print(f"- Macro F1-Score tб»‘t nhбәҘt: {results_by_window[best_w_size]*100:.2f}%")
    
    # TГ¬m Epoch tб»‘t nhбәҘt trung bГ¬nh cho Window Size tб»‘i Ж°u
    best_histories = histories_by_window[best_w_size]
    max_len = max([len(h['val_loss']) for h in best_histories])
    
    avg_loss = np.zeros(max_len)
    avg_val_loss = np.zeros(max_len)
    counts = np.zeros(max_len)
    
    for h in best_histories:
        for idx in range(len(h['loss'])):
            avg_loss[idx] += h['loss'][idx]
            avg_val_loss[idx] += h['val_loss'][idx]
            counts[idx] += 1
            
    avg_loss = np.divide(avg_loss, counts, out=np.zeros_like(avg_loss), where=counts!=0)
    avg_val_loss = np.divide(avg_val_loss, counts, out=np.zeros_like(avg_val_loss), where=counts!=0)
    
    # Epoch tб»‘t nhбәҘt trung bГ¬nh lГ  nЖЎi val_loss trung bГ¬nh Д‘бәЎt cб»ұc tiб»ғu
    best_epoch_avg = np.argmin(avg_val_loss) + 1
    print(f"- Epoch tб»‘t nhбәҘt trung bГ¬nh (dб»ұa trГӘn val_loss cб»ұc tiб»ғu): {best_epoch_avg}")
    print(f"=======================================================")

    # VбәҪ biб»ғu Д‘б»“ trб»ұc quan hГіa
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Biб»ғu Д‘б»“ 1: So sГЎnh Window Size vs F1-score
    sizes = list(results_by_window.keys())
    scores = [results_by_window[s] * 100 for s in sizes]
    bars = ax1.bar([str(s) for s in sizes], scores, color=['#3498db', '#2ecc71', '#e74c3c'], width=0.5)
    ax1.set_title("So sГЎnh Window Size vs Hiб»Үu nДғng F1-Score", fontsize=12, fontweight='bold')
    ax1.set_xlabel("Window Size (Mб»‘c 15s)", fontsize=10)
    ax1.set_ylabel("Macro F1-Score (%)", fontsize=10)
    ax1.set_ylim(0, 100)
    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval:.2f}%", ha='center', va='bottom', fontweight='bold')
    
    # Biб»ғu Д‘б»“ 2: Loss Curve cб»§a Window Size tб»‘i Ж°u Д‘б»ғ giбәЈi thГӯch Early Stopping
    epochs_range = range(1, len(avg_val_loss) + 1)
    ax2.plot(epochs_range, avg_loss, label='Training Loss', color='#2980b9', linewidth=2)
    ax2.plot(epochs_range, avg_val_loss, label='Validation Loss', color='#e67e22', linewidth=2)
    ax2.axvline(x=best_epoch_avg, color='#27ae60', linestyle='--', label=f'Best Epoch: {best_epoch_avg}')
    ax2.scatter(best_epoch_avg, avg_val_loss[best_epoch_avg-1], color='#c0392b', s=100, zorder=5)
    
    ax2.set_title(f"ДҗЖ°б»қng cong Loss cб»§a Window Size tб»‘t nhбәҘt ({best_w_size})", fontsize=12, fontweight='bold')
    ax2.set_xlabel("Epochs", fontsize=10)
    ax2.set_ylabel("Categorical Cross-entropy Loss", fontsize=10)
    ax2.legend()
    ax2.grid(True, linestyle=':', alpha=0.6)
    
    output_img = "hyperparameter_optimization.png"
    plt.tight_layout()
    plt.savefig(output_img, dpi=300)
    print(f"\n[INFO] ДҗГЈ vбәҪ vГ  lЖ°u бәЈnh phГўn tГӯch trб»ұc quan hГіa tбәЎi: {os.path.abspath(output_img)}")
    plt.show()
```

```python
import os
import json
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score
import matplotlib.pyplot as plt
# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')
# KhoбәЈng thб»қi gian tб»« бәЈnh
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"
# Hyperparameter Search Space
WINDOW_SIZES = [16, 32, 64] # CГЎc kГӯch thЖ°б»ӣc cб»ӯa sб»• Д‘б»ғ tГ¬m tб»‘i Ж°u
MAX_EPOCHS = 100            # ChбәЎy tб»‘i Д‘a 100 epochs
BATCH_SIZE = 32
N_CLASSES = 7
# ================= 2. Дҗб»ҠNH NGHДЁA FOLDS (6 - 2 - 2 SPLIT) =================
FOLDS = {
    "fold_1": {"train": [1, 2, 3, 4, 5, 6], "val": [7, 8], "test": [9, 10]},
    "fold_2": {"train": [3, 4, 5, 6, 7, 8], "val": [9, 10], "test": [1, 2]},
    "fold_3": {"train": [5, 6, 7, 8, 9, 10], "val": [1, 2], "test": [3, 4]},
    "fold_4": {"train": [7, 8, 9, 10, 1, 2], "val": [3, 4], "test": [5, 6]},
    "fold_5": {"train": [9, 10, 1, 2, 3, 4], "val": [5, 6], "test": [7, 8]}
}
def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())
START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)
# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE =================
def load_and_align_cow_data(cow_id):
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df
# ================= 4. HГҖM TДӮNG CЖҜб»ңNG Dб»® LIб»ҶU (DATA AUGMENTATION) =================
# ДҗГўy lГ  nЖЎi Д‘б»Ӣnh nghД©a cГЎc phГ©p tДғng cЖ°б»қng dб»Ҝ liб»Үu thГҙ tб»« sensor
def augment_data(X):
    # Jittering: ThГӘm nhiб»…u Gauss ngбә«u nhiГӘn
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    # Scaling: Thay Д‘б»•i ngбә«u nhiГӘn biГӘn Д‘б»ҷ tГӯn hiб»Үu
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    return np.concatenate([X, X_jitter, X_scaled], axis=0)
# ================= 5. SEGMENTATION =================
def segment_cow_data(df, window_size, overlap=0.0):
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list, y_list, w_list = [], [], []
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0)
                w_seg.append(0.0)
            else:
                y_mapped.append(val - 1)
                w_seg.append(1.0)
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
    return np.array(X_list), np.array(y_list), np.array(w_list)
# ================= 6. BILSTM MODEL =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    # BiLSTM sб»ӯ dб»Ҙng 128 units -> Дҗбә§u ra 256 chiб»Ғu trб»ұc tiбәҝp trЖ°б»ӣc phГўn loбәЎi
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model
# ================= 7. HUбәӨN LUYб»ҶN VГҖ TГҢM KIбәҫM WINDOW SIZE Tб»җI ЖҜU =================
results_by_window = {}
histories_by_window = {}
last_X_train_example = None # LЖ°u lбәЎi mб»ҷt mбә«u cб»§a tбәӯp Train Д‘б»ғ vбәҪ minh hб»Қa Augmentation
if __name__ == "__main__":
    print("--- KHб»һI Tбә O CACHE Dб»® LIб»ҶU ---")
    for cid in range(1, 11):
        load_and_align_cow_data(cid)
        
    for w_size in WINDOW_SIZES:
        print(f"\n=======================================================")
        print(f"THб»¬ NGHIб»ҶM WINDOW SIZE = {w_size} (Mб»‘c 15s)")
        print(f"=======================================================")
        
        fold_f1_scores = []
        fold_histories = []
        
        for fold_name, fold_config in FOLDS.items():
            print(f"\n--- {fold_name.upper()} (w_size={w_size}) ---")
            train_ids = fold_config["train"]
            val_ids = fold_config["val"]
            test_ids = fold_config["test"]
            
            def get_data_for_cows(ids, overlap=0.0):
                X_all, y_all, w_all = [], [], []
                for cid in ids:
                    df = load_and_align_cow_data(cid)
                    X_cow, y_cow, w_cow = segment_cow_data(df, w_size, overlap)
                    if len(X_cow) > 0:
                        X_all.append(X_cow)
                        y_all.append(y_cow)
                        w_all.append(w_cow)
                if len(X_all) == 0:
                    return np.empty((0, w_size, 21)), np.empty((0, w_size)), np.empty((0, w_size))
                return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)
            # CбәҜt chuб»—i vб»ӣi overlap = 0.5 (TДғng cЖ°б»қng dб»Ҝ liб»Үu theo trб»Ҙc thб»қi gian)
            X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=0.5)
            X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0)
            X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0)
            
            # LЖ°u mбә«u vГӯ dб»Ҙ trЖ°б»ӣc khi ГЎp dб»Ҙng tДғng cЖ°б»қng dб»Ҝ liб»Үu thГҙ (Д‘б»ғ vбәҪ trб»ұc quan hГіa)
            if last_X_train_example is None and len(X_train) > 0:
                last_X_train_example = X_train[0].copy()
            # Gб»Қi hГ m Augment Д‘б»ғ tДғng gбәҘp 3 lЖ°б»Јng dб»Ҝ liб»Үu (Jittered + Scaled)
            if len(X_train) > 0:
                X_train_augmented = augment_data(X_train)
                y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0)
                w_train_augmented = np.concatenate([w_train, w_train, w_train], axis=0)
                X_train, y_train, w_train = X_train_augmented, y_train_augmented, w_train_augmented
            # Chuбә©n hГіa
            num_samples, num_timesteps, num_features = X_train.shape
            scaler = StandardScaler()
            X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
            X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
            
            if len(X_val) > 0:
                X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], w_size, num_features)
            if len(X_test) > 0:
                X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], w_size, num_features)
                
            y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
            y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
            y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
            
            model = build_bilstm_model(w_size, num_features, N_CLASSES)
            model.compile(
                loss="categorical_crossentropy",
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                metrics=["accuracy"]
            )
            
            # Early stopping kiб»ғm soГЎt tб»‘i Д‘a 100 Epochs. Patience = 8.
            early_stopping = EarlyStopping(
                monitor="val_loss", 
                patience=8, 
                restore_best_weights=True,
                verbose=1
            )
            
            # Fit model
            history = model.fit(
                X_train, y_train_cat,
                sample_weight=w_train,
                epochs=MAX_EPOCHS,
                batch_size=BATCH_SIZE,
                validation_data=(X_val, y_val_cat, w_val),
                callbacks=[early_stopping],
                verbose=1
            )
            
            # ДҗГЎnh giГЎ
            y_pred_cat = model.predict(X_test)
            y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
            y_true = np.argmax(y_test_cat, axis=-1).flatten()
            w_flat = w_test.flatten()
            
            y_pred_valid = y_pred[w_flat == 1.0]
            y_true_valid = y_true[w_flat == 1.0]
            
            fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
            fold_f1_scores.append(fold_f1)
            
            fold_histories.append({
                'loss': history.history['loss'],
                'val_loss': history.history['val_loss']
            })
            
        mean_f1 = np.mean(fold_f1_scores)
        results_by_window[w_size] = mean_f1
        histories_by_window[w_size] = fold_histories
        print(f"\n=> Kбәҝt quбәЈ trung bГ¬nh cho Window Size {w_size}: F1-Score = {mean_f1*100:.2f}%")
    # ================= 8. TГҢM RA THAM Sб»җ Tб»җI ЖҜU NHбәӨT VГҖ Vбәј BIб»ӮU Дҗб»’ =================
    best_w_size = max(results_by_window, key=results_by_window.get)
    print(f"\n=======================================================")
    print(f"KбәҫT QUбәў Tб»җI ЖҜU NHбәӨT:")
    print(f"- Window Size tб»‘t nhбәҘt: {best_w_size}")
    print(f"- Macro F1-Score tб»‘t nhбәҘt: {results_by_window[best_w_size]*100:.2f}%")
    
    # TГ¬m Epoch tб»‘t nhбәҘt trung bГ¬nh cho Window Size tб»‘i Ж°u
    best_histories = histories_by_window[best_w_size]
    max_len = max([len(h['val_loss']) for h in best_histories])
    
    avg_loss = np.zeros(max_len)
    avg_val_loss = np.zeros(max_len)
    counts = np.zeros(max_len)
    
    for h in best_histories:
        for idx in range(len(h['loss'])):
            avg_loss[idx] += h['loss'][idx]
            avg_val_loss[idx] += h['val_loss'][idx]
            counts[idx] += 1
            
    avg_loss = np.divide(avg_loss, counts, out=np.zeros_like(avg_loss), where=counts!=0)
    avg_val_loss = np.divide(avg_val_loss, counts, out=np.zeros_like(avg_val_loss), where=counts!=0)
    
    # Epoch tб»‘t nhбәҘt trung bГ¬nh lГ  nЖЎi val_loss trung bГ¬nh Д‘бәЎt cб»ұc tiб»ғu
    best_epoch_avg = np.argmin(avg_val_loss) + 1
    print(f"- Epoch tб»‘t nhбәҘt trung bГ¬nh (dб»ұa trГӘn val_loss cб»ұc tiб»ғu): {best_epoch_avg}")
    print(f"=======================================================")
    # VбәҪ biб»ғu Д‘б»“ trб»ұc quan hГіa (Bao gб»“m so sГЎnh kГӯch thЖ°б»ӣc cб»ӯa sб»•, Д‘Ж°б»қng cong loss, vГ  trб»ұc quan hГіa augmentation)
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    # 1. So sГЎnh Window Size vs F1-score
    sizes = list(results_by_window.keys())
    scores = [results_by_window[s] * 100 for s in sizes]
    bars = ax1.bar([str(s) for s in sizes], scores, color=['#3498db', '#2ecc71', '#e74c3c'], width=0.5)
    ax1.set_title("1. So sГЎnh Window Size vs Hiб»Үu nДғng F1-Score", fontsize=12, fontweight='bold')
    ax1.set_xlabel("Window Size (Mб»‘c 15s)", fontsize=10)
    ax1.set_ylabel("Macro F1-Score (%)", fontsize=10)
    ax1.set_ylim(0, 100)
    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval:.2f}%", ha='center', va='bottom', fontweight='bold')
    
    # 2. Loss Curve cб»§a Window Size tб»‘i Ж°u Д‘б»ғ giбәЈi thГӯch Early Stopping
    epochs_range = range(1, len(avg_val_loss) + 1)
    ax2.plot(epochs_range, avg_loss, label='Training Loss', color='#2980b9', linewidth=2)
    ax2.plot(epochs_range, avg_val_loss, label='Validation Loss', color='#e67e22', linewidth=2)
    ax2.axvline(x=best_epoch_avg, color='#27ae60', linestyle='--', label=f'Best Epoch: {best_epoch_avg}')
    ax2.scatter(best_epoch_avg, avg_val_loss[best_epoch_avg-1], color='#c0392b', s=100, zorder=5)
    
    ax2.set_title(f"2. ДҗЖ°б»қng cong Loss cб»§a Window Size tб»‘t nhбәҘt ({best_w_size})", fontsize=12, fontweight='bold')
    ax2.set_xlabel("Epochs", fontsize=10)
    ax2.set_ylabel("Categorical Cross-entropy Loss", fontsize=10)
    ax2.legend()
    ax2.grid(True, linestyle=':', alpha=0.6)
    # Trб»ұc quan hГіa dб»Ҝ liб»Үu gб»‘c vs Jittered & Scaled (Sб»ӯ dб»Ҙng 1 mбә«u vГӯ dб»Ҙ б»ҹ trб»Ҙc 1 - immu_accel_y)
    if last_X_train_example is not None:
        # Chб»Қn Д‘бә·c trЖ°ng thб»© 4 (Gia tб»‘c Y cб»§a IMMU)
        feature_idx = 4
        # LбәҘy nhГЈn Д‘бә·c trЖ°ng
        feature_name = "IMMU Accel_Y"
        
        orig_sig = last_X_train_example[:, feature_idx]
        time_steps = np.arange(len(orig_sig))
        
        # 3. Trб»ұc quan hГіa Jittering (ThГӘm nhiб»…u)
        noise = np.random.normal(0, 0.05, size=orig_sig.shape)
        jitter_sig = orig_sig + noise
        ax3.plot(time_steps, orig_sig, label='Dб»Ҝ liб»Үu gб»‘c (Original)', color='#2c3e50', linewidth=2)
        ax3.plot(time_steps, jitter_sig, label='TДғng cЖ°б»қng Jittering (+Nhiб»…u Gauss)', color='#e74c3c', linestyle=':', linewidth=1.5)
        ax3.set_title(f"3. Trб»ұc quan hГіa phГ©p Augment Jittering ({feature_name})", fontsize=12, fontweight='bold')
        ax3.set_xlabel("Time step (15s)", fontsize=10)
        ax3.set_ylabel("GiГЎ trб»Ӣ Д‘бә·c trЖ°ng", fontsize=10)
        ax3.legend()
        ax3.grid(True, linestyle=':', alpha=0.6)
        
        # 4. Trб»ұc quan hГіa Scaling (Biбәҝn Д‘б»•i biГӘn Д‘б»ҷ)
        scale_factor = 1.25
        scale_sig = orig_sig * scale_factor
        ax4.plot(time_steps, orig_sig, label='Dб»Ҝ liб»Үu gб»‘c (Original)', color='#2c3e50', linewidth=2)
        ax4.plot(time_steps, scale_sig, label=f'TДғng cЖ°б»қng Scaling (Scale x{scale_factor})', color='#9b59b6', linestyle='--', linewidth=1.5)
        ax4.set_title(f"4. Trб»ұc quan hГіa phГ©p Augment Scaling ({feature_name})", fontsize=12, fontweight='bold')
        ax4.set_xlabel("Time step (15s)", fontsize=10)
        ax4.set_ylabel("GiГЎ trб»Ӣ Д‘бә·c trЖ°ng", fontsize=10)
        ax4.legend()
        ax4.grid(True, linestyle=':', alpha=0.6)
        
    output_img = "hyperparameter_optimization.png"
    plt.tight_layout()
    plt.savefig(output_img, dpi=300)
    print(f"\n[INFO] ДҗГЈ vбәҪ vГ  lЖ°u бәЈnh phГўn tГӯch trб»ұc quan hГіa tбәЎi: {os.path.abspath(output_img)}")
    plt.show()
```

```python

import os
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

# KhoбәЈng thб»қi gian tб»« dб»Ҝ liб»Үu thб»ұc tбәҝ
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

# Tham sб»‘ mГҙ hГ¬nh & huбәҘn luyб»Үn tб»‘i Ж°u
WINDOW_SIZE = 16       # Дҗб»ҷ dГ i chuб»—i (sб»‘ bЖ°б»ӣc 15s liГӘn tiбәҝp Д‘Ж°a vГ o BiLSTM)
N_CLASSES = 7          # 7 hГ nh vi chГӯnh (bб»Ҹ behavior 0)
EPOCHS = 100           # TДғng tб»‘i Д‘a 100 epochs Д‘б»ғ khб»ӣp cбәҘu hГ¬nh tб»‘t nhбәҘt
BATCH_SIZE = 32

# ================= 2. Дҗб»ҠNH NGHДЁA 5-FOLD (CHIA THEO CON BГ’ 6 - 2 - 2) =================
FOLDS = {
    "fold_1": {"train": [1, 2, 3, 4, 5, 6], "val": [7, 8], "test": [9, 10]},
    "fold_2": {"train": [3, 4, 5, 6, 7, 8], "val": [9, 10], "test": [1, 2]},
    "fold_3": {"train": [5, 6, 7, 8, 9, 10], "val": [1, 2], "test": [3, 4]},
    "fold_4": {"train": [7, 8, 9, 10, 1, 2], "val": [3, 4], "test": [5, 6]},
    "fold_5": {"train": [9, 10, 1, 2, 3, 4], "val": [5, 6], "test": [7, 8]}
}

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE CHO Tб»ӘNG CON BГ’ =================
def load_and_align_cow_data(cow_id):
    """Дҗб»Қc vГ  gб»ҷp tбәҘt cбәЈ sensor cho 1 con bГІ, Д‘б»“ng bб»ҷ vб»Ғ lЖ°б»ӣi 15s"""
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    # Дҗб»Қc 6 sensor vГ  nhГЈn behavior
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    # LЖ°б»ӣi thб»қi gian 15s chuбә©n tб»« Start -> End
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    # HГ m merge
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    # Дҗб»“ng bб»ҷ hГіa cГЎc kГӘnh cбәЈm biбәҝn
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    # Дҗiб»Ғn giГЎ trб»Ӣ trб»‘ng
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# TбәЎo sбәөn cache cho cбәЈ 10 con bГІ
print("--- KHб»һI Tбә O BЖҜб»ҡC Дҗб»’NG Bб»ҳ VГҖ CACHE Dб»® LIб»ҶU ---")
for cid in range(1, 11):
    load_and_align_cow_data(cid)

# ================= 4. HГҖM TДӮNG CЖҜб»ңNG Dб»® LIб»ҶU (DATA AUGMENTATION) =================
def augment_data(X):
    """
    ГҒp dб»Ҙng 2 kб»№ thuбәӯt Augmentation phб»• biбәҝn cho dб»Ҝ liб»Үu cбәЈm biбәҝn:
    1. Jittering: ThГӘm nhiб»…u Gaussian ngбә«u nhiГӘn
    2. Scaling: NhГўn tб»· lб»Ү ngбә«u nhiГӘn cho biГӘn Д‘б»ҷ tГӯn hiб»Үu
    """
    # Jittering
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    
    # Scaling
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    
    # TrбәЈ vб»Ғ gб»ҷp cбәЈ dб»Ҝ liб»Үu gб»‘c vГ  dб»Ҝ liб»Үu Д‘ГЈ tДғng cЖ°б»қng
    return np.concatenate([X, X_jitter, X_scaled], axis=0)

# ================= 5. PHГӮN ДҗOбә N CHUб»–I (SEGMENTATION & WINDOW SHIFT) =================
def segment_cow_data(df, window_size, overlap=0.0):
    """CбәҜt dб»Ҝ liб»Үu chuб»—i dГ i cб»§a bГІ thГ nh cГЎc Д‘oбәЎn window_size bЖ°б»ӣc"""
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list = []
    y_list = []
    w_list = [] # sample weights Д‘б»ғ bб»Ҹ qua nhГЈn 0 khi tГӯnh Loss
    
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0) # dummy class
                w_seg.append(0.0)  # ignore loss (trб»Қng sб»‘ 0)
            else:
                y_mapped.append(val - 1) # chuyб»ғn Д‘б»•i nhГЈn 1..7 vб»Ғ 0..6
                w_seg.append(1.0)        # tГӯnh loss bГ¬nh thЖ°б»қng
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
        
    return np.array(X_list), np.array(y_list), np.array(w_list)

# ================= 6. KHб»һI Tбә O MГ” HГҢNH SENSOR BILSTM ДҗбәҰU RA 256 ДҗГғ LЖҜб»ўC Bб»Һ FC2 =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    
    # 1. NhГЎnh BiLSTM cб»§a Sensor: 128 units -> Дҗбә§u ra trб»ұc tiбәҝp lГ  256 chiб»Ғu
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    # ГҒp dб»Ҙng trб»ұc tiбәҝp Regularization lГӘn Д‘бә§u ra BiLSTM Д‘б»ғ train б»•n Д‘б»Ӣnh
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    
    # Lб»ӣp phГўn loбәЎi Sequence-to-Sequence (DГ№ng Д‘б»ғ train Д‘б»ҷc lбәӯp nhГЎnh sensor)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    
    # TбәЎo model hoГ n chб»үnh Д‘б»ғ huбәҘn luyб»Үn Д‘б»ҷc lбәӯp
    model = Model(inputs=inputs, outputs=outputs, name="Sensor_BiLSTM_Classifier")
    
    # Дҗб»“ng thб»қi trГӯch xuбәҘt riГӘng Feature Extractor (Д‘бә§u ra shape 256 cб»§a lб»ӣp BiLSTM_Core)
    feature_extractor = Model(inputs=inputs, outputs=lstm_out, name="Sensor_Feature_Extractor_256")
    
    return model, feature_extractor

# ================= 7. VГ’NG HUбәӨN LUYб»ҶN 5-FOLD CROSS-VALIDATION =================
aggregated_f1 = []
best_epochs = {}

for fold_name, fold_config in FOLDS.items():
    print(f"\n==================== HUбәӨN LUYб»ҶN {fold_name.upper()} ====================")
    train_ids = fold_config["train"]
    val_ids = fold_config["val"]
    test_ids = fold_config["test"]
    
    print(f"Train Cows: {train_ids} (6 con)")
    print(f"Val Cows: {val_ids} (2 con)")
    print(f"Test Cows: {test_ids} (2 con)")
    
    # Load vГ  phГўn Д‘oбәЎn dб»Ҝ liб»Үu
    def get_data_for_cows(ids, overlap=0.0):
        X_all, y_all, w_all = [], [], []
        for cid in ids:
            df = load_and_align_cow_data(cid)
            X_cow, y_cow, w_cow = segment_cow_data(df, WINDOW_SIZE, overlap)
            if len(X_cow) > 0:
                X_all.append(X_cow)
                y_all.append(y_cow)
                w_all.append(w_cow)
        return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

    # Train: sб»ӯ dб»Ҙng overlap 50% Д‘б»ғ sinh nhiб»Ғu mбә«u hЖЎn (TДғng cЖ°б»қng dб»Ҝ liб»Үu thб»қi gian)
    X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=0.5)
    # Val/Test: sб»ӯ dб»Ҙng overlap 0% Д‘б»ғ Д‘ГЎnh giГЎ cГҙng bбәұng
    X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0)
    X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0)
    
    # ГҒp dб»Ҙng Augmentation (Jittering + Scaling) cho tбәӯp Train
    if len(X_train) > 0:
        X_train_augmented = augment_data(X_train)
        y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0)
        w_train_augmented = np.concatenate([w_train, w_train, w_train], axis=0)
        print(f"Sб»‘ lЖ°б»Јng mбә«u Train sau khi Augment: {X_train.shape[0]} -> {X_train_augmented.shape[0]}")
        X_train, y_train, w_train = X_train_augmented, y_train_augmented, w_train_augmented

    # Chuбә©n hГіa (Fit scaler trГӘn tбәӯp Train, Transform trГӘn Val/Test)
    num_samples, num_timesteps, num_features = X_train.shape
    scaler = StandardScaler()
    X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
    X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
    
    if len(X_val) > 0:
        X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], WINDOW_SIZE, num_features)
    if len(X_test) > 0:
        X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], WINDOW_SIZE, num_features)
        
    # Chuyб»ғn Д‘б»•i nhГЈn sang One-hot
    y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
    y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
    y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
    
    # Khб»ҹi tбәЎo mГҙ hГ¬nh
    model, extractor = build_bilstm_model(WINDOW_SIZE, num_features, N_CLASSES)
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics=["accuracy"]
    )
    
    # Callbacks (Patience = 8 Д‘б»ғ EarlyStopping tб»‘i Ж°u viб»Үc tГ¬m Д‘iб»ғm dб»«ng tб»‘t nhбәҘt)
    checkpoint_path = f"best_model_{fold_name}.h5"
    early_stopping = EarlyStopping(
        monitor="val_loss", 
        patience=8, 
        restore_best_weights=True,
        verbose=1
    )
    callbacks = [
        early_stopping,
        ModelCheckpoint(checkpoint_path, monitor="val_loss", save_best_only=True)
    ]
    
    # Fit model (truyб»Ғn sample_weight Д‘б»ғ bб»Ҹ qua nhГЈn 0 khi tГӯnh Loss)
    history = model.fit(
        X_train, y_train_cat,
        sample_weight=w_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_val, y_val_cat, w_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # TГ¬m ra epoch tб»‘t nhбәҘt tб»« lб»Ӣch sб»ӯ val_loss cб»§a EarlyStopping
    best_epoch = np.argmin(history.history['val_loss']) + 1
    best_val_loss = history.history['val_loss'][best_epoch - 1]
    best_epochs[fold_name] = (best_epoch, best_val_loss)
    print(f"\n[INFO] {fold_name.upper()} dб»«ng lбәЎi. Epoch tб»‘t nhбәҘt Д‘Ж°б»Јc lЖ°u lбәЎi: Epoch {best_epoch} (Val Loss tб»‘t nhбәҘt: {best_val_loss:.4f})")
    
    # ДҗГЎnh giГЎ trГӘn tбәӯp test Д‘б»ҷc lбәӯp cб»§a fold
    y_pred_cat = model.predict(X_test)
    y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
    y_true = np.argmax(y_test_cat, axis=-1).flatten()
    w_flat = w_test.flatten()
    
    # Chб»ү Д‘ГЎnh giГЎ trГӘn cГЎc bЖ°б»ӣc thб»қi gian cГі nhГЈn hб»Јp lб»Ү (w == 1.0)
    y_pred_valid = y_pred[w_flat == 1.0]
    y_true_valid = y_true[w_flat == 1.0]
    
    fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
    fold_acc = accuracy_score(y_true_valid, y_pred_valid)
    aggregated_f1.append(fold_f1)
    
    print(f"KбәҫT QUбәў Dб»° ДҗOГҒN {fold_name.upper()}:")
    print(f"- Accuracy (chб»ү tГӯnh nhГЈn hб»Јp lб»Ү): {fold_acc:.4f}")
    print(f"- Macro F1-Score (chб»ү tГӯnh nhГЈn hб»Јp lб»Ү): {fold_f1:.4f}")

# ================= 8. IN BбәўNG Tб»”NG Hб»ўP KбәҫT QUбәў VГҖ OUTPUT Cб»ҰA BILSTM =================
print("\n=======================================================")
print("============= THГ”NG TIN CГҒC EPOCH Tб»җT NHбәӨT =============")
print("=======================================================")
for fold_name, (epoch, loss) in best_epochs.items():
    print(f"- {fold_name.upper()}: Epoch tб»‘t nhбәҘt = {epoch} | Val Loss = {loss:.4f}")

print("\n==================== KбәҫT QUбәў CHUNG 5-FOLD ====================")
print(f"Macro F1 trung bГ¬nh: {np.mean(aggregated_f1)*100:.2f}% Вұ {np.std(aggregated_f1)*100:.2f}%")

print("\n=================== KHбәўO SГҒT OUTPUT Cб»ҰA SENSOR BILSTM (256 Chiб»Ғu) ===================")
# Sб»ӯ dб»Ҙng extractor cб»§a fold cuб»‘i cГ№ng Д‘б»ғ demo trГӯch xuбәҘt Д‘бә·c trЖ°ng
if len(X_test) > 0:
    test_features = extractor.predict(X_test)
    print(f"- Input shape cб»§a Sensor (Samples, Window_size, Sensors_features): {X_test.shape}")
    print(f"- Output shape trГӯch xuбәҘt tб»« BiLSTM (Samples, Window_size, 256): {test_features.shape}")
    print(f"- GiГЎ trб»Ӣ Д‘бә·c trЖ°ng demo cб»§a mбә«u Д‘бә§u tiГӘn tбәЎi timestep Д‘бә§u tiГӘn (10 giГЎ trб»Ӣ Д‘бә§u):")
    print(test_features[0, 0, :10])
    print(f"- Model trГӯch xuбәҘt Д‘бә·c trЖ°ng cГі cбәҘu trГәc inputs: {extractor.input_names} | outputs: {extractor.output_names}")
```

```python
import os
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

# KhoбәЈng thб»қi gian tб»« dб»Ҝ liб»Үu thб»ұc tбәҝ
START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

# Tham sб»‘ mГҙ hГ¬nh & huбәҘn luyб»Үn tб»‘i Ж°u
WINDOW_SIZE = 16       # Дҗб»ҷ dГ i chuб»—i (sб»‘ bЖ°б»ӣc 15s liГӘn tiбәҝp Д‘Ж°a vГ o BiLSTM)
N_CLASSES = 7          # 7 hГ nh vi chГӯnh (bб»Ҹ behavior 0)
EPOCHS = 100           # TДғng tб»‘i Д‘a 100 epochs Д‘б»ғ khб»ӣp cбәҘu hГ¬nh tб»‘t nhбәҘt
BATCH_SIZE = 32

# ================= 2. Дҗб»ҠNH NGHДЁA 5-FOLD (CHIA THEO CON BГ’ 6 - 2 - 2) =================
FOLDS = {
    "fold_1": {"train": [1, 2, 3, 4, 5, 6], "val": [7, 8], "test": [9, 10]},
    "fold_2": {"train": [3, 4, 5, 6, 7, 8], "val": [9, 10], "test": [1, 2]},
    "fold_3": {"train": [5, 6, 7, 8, 9, 10], "val": [1, 2], "test": [3, 4]},
    "fold_4": {"train": [7, 8, 9, 10, 1, 2], "val": [3, 4], "test": [5, 6]},
    "fold_5": {"train": [9, 10, 1, 2, 3, 4], "val": [5, 6], "test": [7, 8]}
}

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 3. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE CHO Tб»ӘNG CON BГ’ =================
def load_and_align_cow_data(cow_id):
    """Дҗб»Қc vГ  gб»ҷp tбәҘt cбәЈ sensor cho 1 con bГІ, Д‘б»“ng bб»ҷ vб»Ғ lЖ°б»ӣi 15s"""
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    # Дҗб»Қc 6 sensor vГ  nhГЈn behavior
    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    # LЖ°б»ӣi thб»қi gian 15s chuбә©n tб»« Start -> End
    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    # HГ m merge
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    # Дҗб»“ng bб»ҷ hГіa cГЎc kГӘnh cбәЈm biбәҝn
    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    # Дҗiб»Ғn giГЎ trб»Ӣ trб»‘ng
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# TбәЎo sбәөn cache cho cбәЈ 10 con bГІ
print("--- KHб»һI Tбә O BЖҜб»ҡC Дҗб»’NG Bб»ҳ VГҖ CACHE Dб»® LIб»ҶU ---")
for cid in range(1, 11):
    load_and_align_cow_data(cid)

# ================= 4. HГҖM TДӮNG CЖҜб»ңNG Dб»® LIб»ҶU (DATA AUGMENTATION) =================
def augment_data(X):
    """
    ГҒp dб»Ҙng 2 kб»№ thuбәӯt Augmentation phб»• biбәҝn cho dб»Ҝ liб»Үu cбәЈm biбәҝn:
    1. Jittering: ThГӘm nhiб»…u Gaussian ngбә«u nhiГӘn
    2. Scaling: NhГўn tб»· lб»Ү ngбә«u nhiГӘn cho biГӘn Д‘б»ҷ tГӯn hiб»Үu
    """
    # Jittering
    noise = np.random.normal(0, 0.03, size=X.shape)
    X_jitter = X + noise
    
    # Scaling
    factor = np.random.normal(1.0, 0.1, size=(X.shape[0], 1, X.shape[2]))
    X_scaled = X * factor
    
    # TrбәЈ vб»Ғ gб»ҷp cбәЈ dб»Ҝ liб»Үu gб»‘c vГ  dб»Ҝ liб»Үu Д‘ГЈ tДғng cЖ°б»қng
    return np.concatenate([X, X_jitter, X_scaled], axis=0)

# ================= 5. PHГӮN ДҗOбә N CHUб»–I (SEGMENTATION & WINDOW SHIFT) =================
def segment_cow_data(df, window_size, overlap=0.0, use_class_weights=False):
    """CбәҜt dб»Ҝ liб»Үu chuб»—i dГ i cб»§a bГІ thГ nh cГЎc Д‘oбәЎn window_size bЖ°б»ӣc"""
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    
    # Дҗб»Ӣnh nghД©a trб»Қng sб»‘ lб»ӣp tГӯnh sбәөn tб»« phГўn phб»‘i dб»Ҝ liб»Үu thб»ұc tбәҝ:
    # W_c = Tб»•ng mбә«u hб»Јp lб»Ү (36903) / (7 * Sб»‘ mбә«u lб»ӣp c)
    CLASS_WEIGHTS = {
        0: 10.43, # NhГЈn 1 (Feeding) -> Index 0
        1: 0.56,  # NhГЈn 2 (Lying) -> Index 1
        2: 2.24,  # NhГЈn 3 (Standing)
        3: 1.39,  # NhГЈn 4 (Walking)
        4: 20.27, # NhГЈn 5 (Drinking)
        5: 11.43, # NhГЈn 6 (Grooming)
        6: 0.26   # NhГЈn 7 (Restless)
    }
    
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list = []
    y_list = []
    w_list = [] # sample weights Д‘б»ғ bб»Ҹ qua nhГЈn 0 khi tГӯnh Loss
    
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0) # dummy class
                w_seg.append(0.0)  # ignore loss (trб»Қng sб»‘ 0)
            else:
                mapped_val = val - 1
                y_mapped.append(mapped_val) # chuyб»ғn Д‘б»•i nhГЈn 1..7 vб»Ғ 0..6
                if use_class_weights:
                    w_seg.append(CLASS_WEIGHTS[mapped_val])
                else:
                    w_seg.append(1.0)        # tГӯnh loss bГ¬nh thЖ°б»қng
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return np.empty((0, window_size, features.shape[1])), np.empty((0, window_size)), np.empty((0, window_size))
        
    return np.array(X_list), np.array(y_list), np.array(w_list)

# ================= 6. KHб»һI Tбә O MГ” HГҢNH SENSOR BILSTM ДҗбәҰU RA 256 ДҗГғ LЖҜб»ўC Bб»Һ FC2 =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    
    # 1. NhГЎnh BiLSTM cб»§a Sensor: 128 units -> Дҗбә§u ra trб»ұc tiбәҝp lГ  256 chiб»Ғu
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    
    # ГҒp dб»Ҙng trб»ұc tiбәҝp Regularization lГӘn Д‘бә§u ra BiLSTM Д‘б»ғ train б»•n Д‘б»Ӣnh
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    
    # Lб»ӣp phГўn loбәЎi Sequence-to-Sequence (DГ№ng Д‘б»ғ train Д‘б»ҷc lбәӯp nhГЎnh sensor)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    
    # TбәЎo model hoГ n chб»үnh Д‘б»ғ huбәҘn luyб»Үn Д‘б»ҷc lбәӯp
    model = Model(inputs=inputs, outputs=outputs, name="Sensor_BiLSTM_Classifier")
    
    # Дҗб»“ng thб»қi trГӯch xuбәҘt riГӘng Feature Extractor (Д‘бә§u ra shape 256 cб»§a lб»ӣp BiLSTM_Core)
    feature_extractor = Model(inputs=inputs, outputs=lstm_out, name="Sensor_Feature_Extractor_256")
    
    return model, feature_extractor

# ================= 7. VГ’NG HUбәӨN LUYб»ҶN 5-FOLD CROSS-VALIDATION =================
aggregated_f1 = []
best_epochs = {}

for fold_name, fold_config in FOLDS.items():
    print(f"\n==================== HUбәӨN LUYб»ҶN {fold_name.upper()} ====================")
    train_ids = fold_config["train"]
    val_ids = fold_config["val"]
    test_ids = fold_config["test"]
    
    print(f"Train Cows: {train_ids} (6 con)")
    print(f"Val Cows: {val_ids} (2 con)")
    print(f"Test Cows: {test_ids} (2 con)")
    
    # Load vГ  phГўn Д‘oбәЎn dб»Ҝ liб»Үu
    def get_data_for_cows(ids, overlap=0.0, use_class_weights=False):
        X_all, y_all, w_all = [], [], []
        for cid in ids:
            df = load_and_align_cow_data(cid)
            X_cow, y_cow, w_cow = segment_cow_data(df, WINDOW_SIZE, overlap, use_class_weights)
            if len(X_cow) > 0:
                X_all.append(X_cow)
                y_all.append(y_cow)
                w_all.append(w_cow)
        return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

    # Train: sб»ӯ dб»Ҙng overlap 50% vГ  ГЎp dб»Ҙng Class Weights Д‘б»ғ cГўn bбәұng loss
    X_train, y_train, w_train = get_data_for_cows(train_ids, overlap=0.5, use_class_weights=True)
    # Val/Test: sб»ӯ dб»Ҙng overlap 0% Д‘б»ғ Д‘ГЎnh giГЎ cГҙng bбәұng
    X_val, y_val, w_val = get_data_for_cows(val_ids, overlap=0.0, use_class_weights=False)
    X_test, y_test, w_test = get_data_for_cows(test_ids, overlap=0.0, use_class_weights=False)
    
    # ГҒp dб»Ҙng Augmentation (Jittering + Scaling) cho tбәӯp Train
    if len(X_train) > 0:
        X_train_augmented = augment_data(X_train)
        y_train_augmented = np.concatenate([y_train, y_train, y_train], axis=0)
        w_train_augmented = np.concatenate([w_train, w_train, w_train], axis=0)
        print(f"Sб»‘ lЖ°б»Јng mбә«u Train sau khi Augment: {X_train.shape[0]} -> {X_train_augmented.shape[0]}")
        X_train, y_train, w_train = X_train_augmented, y_train_augmented, w_train_augmented

    # Chuбә©n hГіa (Fit scaler trГӘn tбәӯp Train, Transform trГӘn Val/Test)
    num_samples, num_timesteps, num_features = X_train.shape
    scaler = StandardScaler()
    X_train_2d = scaler.fit_transform(X_train.reshape(-1, num_features))
    X_train = X_train_2d.reshape(num_samples, num_timesteps, num_features)
    
    if len(X_val) > 0:
        X_val = scaler.transform(X_val.reshape(-1, num_features)).reshape(X_val.shape[0], WINDOW_SIZE, num_features)
    if len(X_test) > 0:
        X_test = scaler.transform(X_test.reshape(-1, num_features)).reshape(X_test.shape[0], WINDOW_SIZE, num_features)
        
    # Chuyб»ғn Д‘б»•i nhГЈn sang One-hot
    y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
    y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
    y_test_cat = to_categorical(y_test, num_classes=N_CLASSES)
    
    # Khб»ҹi tбәЎo mГҙ hГ¬nh
    model, extractor = build_bilstm_model(WINDOW_SIZE, num_features, N_CLASSES)
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics=["accuracy"]
    )
    
    # Callbacks (Patience = 8 Д‘б»ғ EarlyStopping tб»‘i Ж°u viб»Үc tГ¬m Д‘iб»ғm dб»«ng tб»‘t nhбәҘt)
    checkpoint_path = f"best_model_{fold_name}.h5"
    early_stopping = EarlyStopping(
        monitor="val_loss", 
        patience=8, 
        restore_best_weights=True,
        verbose=1
    )
    callbacks = [
        early_stopping,
        ModelCheckpoint(checkpoint_path, monitor="val_loss", save_best_only=True)
    ]
    
    # Fit model (truyб»Ғn sample_weight chб»©a Class Weights)
    history = model.fit(
        X_train, y_train_cat,
        sample_weight=w_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_val, y_val_cat, w_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # TГ¬m ra epoch tб»‘t nhбәҘt tб»« lб»Ӣch sб»ӯ val_loss cб»§a EarlyStopping
    best_epoch = np.argmin(history.history['val_loss']) + 1
    best_val_loss = history.history['val_loss'][best_epoch - 1]
    best_epochs[fold_name] = (best_epoch, best_val_loss)
    print(f"\n[INFO] {fold_name.upper()} dб»«ng lбәЎi. Epoch tб»‘t nhбәҘt Д‘Ж°б»Јc lЖ°u lбәЎi: Epoch {best_epoch} (Val Loss tб»‘t nhбәҘt: {best_val_loss:.4f})")
    
    # ДҗГЎnh giГЎ trГӘn tбәӯp test Д‘б»ҷc lбәӯp cб»§a fold
    y_pred_cat = model.predict(X_test)
    y_pred = np.argmax(y_pred_cat, axis=-1).flatten()
    y_true = np.argmax(y_test_cat, axis=-1).flatten()
    w_flat = w_test.flatten()
    
    # Chб»ү Д‘ГЎnh giГЎ trГӘn cГЎc bЖ°б»ӣc thб»қi gian cГі nhГЈn hб»Јp lб»Ү (w == 1.0)
    y_pred_valid = y_pred[w_flat == 1.0]
    y_true_valid = y_true[w_flat == 1.0]
    
    fold_f1 = f1_score(y_true_valid, y_pred_valid, average="macro")
    fold_acc = accuracy_score(y_true_valid, y_pred_valid)
    aggregated_f1.append(fold_f1)
    
    print(f"KбәҫT QUбәў Dб»° ДҗOГҒN {fold_name.upper()}:")
    print(f"- Accuracy (chб»ү tГӯnh nhГЈn hб»Јp lб»Ү): {fold_acc:.4f}")
    print(f"- Macro F1-Score (chб»ү tГӯnh nhГЈn hб»Јp lб»Ү): {fold_f1:.4f}")

# ================= 8. IN BбәўNG Tб»”NG Hб»ўP KбәҫT QUбәў VГҖ OUTPUT Cб»ҰA BILSTM =================
print("\n=======================================================")
print("============= THГ”NG TIN CГҒC EPOCH Tб»җT NHбәӨT =============")
print("=======================================================")
for fold_name, (epoch, loss) in best_epochs.items():
    print(f"- {fold_name.upper()}: Epoch tб»‘t nhбәҘt = {epoch} | Val Loss = {loss:.4f}")

print("\n==================== KбәҫT QUбәў CHUNG 5-FOLD ====================")
print(f"Macro F1 trung bГ¬nh: {np.mean(aggregated_f1)*100:.2f}% Вұ {np.std(aggregated_f1)*100:.2f}%")

print("\n=================== KHбәўO SГҒT OUTPUT Cб»ҰA SENSOR BILSTM (256 Chiб»Ғu) ===================")
# Sб»ӯ dб»Ҙng extractor cб»§a fold cuб»‘i cГ№ng Д‘б»ғ demo trГӯch xuбәҘt Д‘бә·c trЖ°ng
if len(X_test) > 0:
    test_features = extractor.predict(X_test)
    print(f"- Input shape cб»§a Sensor (Samples, Window_size, Sensors_features): {X_test.shape}")
    print(f"- Output shape trГӯch xuбәҘt tб»« BiLSTM (Samples, Window_size, 256): {test_features.shape}")
    print(f"- GiГЎ trб»Ӣ Д‘бә·c trЖ°ng demo cб»§a mбә«u Д‘бә§u tiГӘn tбәЎi timestep Д‘бә§u tiГӘn (10 giГЎ trб»Ӣ Д‘бә§u):")
    print(test_features[0, 0, :10])
    print(f"- CбәҘu trГәc Model trГӯch xuбәҘt Д‘бә·c trЖ°ng:")
    print(f"  + Input shape: {extractor.input_shape}")
    print(f"  + Output shape: {extractor.output_shape}")
```

```python
import os
import numpy as np
import pandas as pd
import pytz
from datetime import datetime
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Dense, Bidirectional, LSTM, BatchNormalization, Dropout, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score

# ================= 1. THГ”NG Sб»җ CбәӨU HГҢNH =================
SENSOR_DATA_DIR = r"d:\archive\sensor_data\sensor_data"
CACHE_DIR = r"d:\archive\sensor_data\cache"
DATES = ["0725", "0726"]
CT_TIME_ZONE = pytz.timezone('America/Chicago')

START_DT = "2023-07-25 07:57:26"
END_DT = "2023-07-26 04:57:11"

WINDOW_SIZE = 16       # KГӯch thЖ°б»ӣc cб»ӯa sб»• tб»‘i Ж°u (mб»‘c 15s)
N_FEATURES = 20        # 20 cбәЈm biбәҝn Д‘бә§u vГ o
N_CLASSES = 7          # 7 hГ nh vi chГӯnh
BATCH_SIZE = 32
MAX_EPOCHS = 100

# Chia Fold 2: Train [3-8], Val [9-10], Test [1-2]
TRAIN_COWS = [3, 4, 5, 6, 7, 8]
VAL_COWS = [9, 10]
TEST_COWS = [1, 2]

MODEL_PATH = "best_model_no_aug_fold_2.h5"

def datetime_to_unix(dt_str):
    dt = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    dt = CT_TIME_ZONE.localize(dt)
    return int(dt.timestamp())

START_TS = datetime_to_unix(START_DT)
END_TS = datetime_to_unix(END_DT)

# ================= 2. Дҗб»’NG Bб»ҳ VГҖ LЖҜU CACHE (NбәҫU CHЖҜA CГ“) =================
def load_and_align_cow_data(cow_id):
    tag_name = f"T{cow_id:02d}"
    cow_name = f"C{cow_id:02d}"
    os.makedirs(CACHE_DIR, exist_ok=True)
    cache_path = os.path.join(CACHE_DIR, f"{cow_name}_aligned_15s.csv")
    
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)
    
    print(f"Дҗang Д‘б»“ng bб»ҷ dб»Ҝ liб»Үu cho bГІ {cow_name}...")
    
    def load_sensor(sensor_type, is_sub_data=False, use_cow_name=False, has_dates=True):
        combined = pd.DataFrame()
        folder = "sub_data" if is_sub_data else "main_data"
        name = cow_name if use_cow_name else tag_name
        
        if has_dates:
            for date in DATES:
                filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, name, f"{name}_{date}.csv")
                if os.path.exists(filepath):
                    combined = pd.concat([combined, pd.read_csv(filepath)], ignore_index=True)
        else:
            filepath = os.path.join(SENSOR_DATA_DIR, folder, sensor_type, f"{name}.csv")
            if os.path.exists(filepath):
                combined = pd.read_csv(filepath)
                
        if not combined.empty:
            combined['timestamp'] = combined['timestamp'].astype(np.float64)
            combined = combined.sort_values('timestamp').reset_index(drop=True)
        return combined

    uwb = load_sensor("uwb")
    immu = load_sensor("immu")
    ankle = load_sensor("ankle", use_cow_name=True)
    ankle_accel = load_sensor("ankle_accel", is_sub_data=True, use_cow_name=True, has_dates=False)
    head = load_sensor("head_direction", is_sub_data=True)
    pressure = load_sensor("pressure")
    
    labels_df = pd.DataFrame()
    for date in DATES:
        lbl_path = os.path.join(SENSOR_DATA_DIR, "behavior_labels", "individual", f"{cow_name}_{date}.csv")
        if os.path.exists(lbl_path):
            labels_df = pd.concat([labels_df, pd.read_csv(lbl_path)], ignore_index=True)
    if not labels_df.empty:
        labels_df['timestamp'] = labels_df['timestamp'].astype(np.float64)
        labels_df = labels_df.sort_values('timestamp').reset_index(drop=True)

    grid_timestamps = np.arange(START_TS, END_TS + 1, 15).astype(np.float64)
    master_df = pd.DataFrame({'timestamp': grid_timestamps})
    
    def merge_sensor(main_df, sensor_df, cols, prefix):
        if not sensor_df.empty:
            existing = [c for c in cols if c in sensor_df.columns]
            sub = sensor_df[existing].copy()
            rename = {c: f"{prefix}_{c}" for c in existing if c != 'timestamp'}
            sub = sub.rename(columns=rename)
            main_df = pd.merge_asof(main_df, sub, on='timestamp', direction='nearest', tolerance=7.5)
        return main_df

    master_df = merge_sensor(master_df, uwb, ['timestamp', 'coord_x_cm', 'coord_y_cm', 'coord_z_cm'], "uwb")
    master_df = merge_sensor(master_df, immu, ['timestamp', 'accel_x_mps2', 'accel_y_mps2', 'accel_z_mps2', 'mag_x_uT', 'mag_y_uT', 'mag_z_uT'], "immu")
    master_df = merge_sensor(master_df, ankle, ['timestamp', 'lying'], "ankle")
    master_df = merge_sensor(master_df, ankle_accel, ['timestamp', 'accel_x', 'accel_y', 'accel_z'], "ankle_accel")
    master_df = merge_sensor(master_df, head, ['timestamp', 'roll', 'pitch', 'yaw', 'accel_norm', 'relative_angle'], "head")
    master_df = merge_sensor(master_df, pressure, ['timestamp', 'pressure_Pa', 'elevation_m'], "pressure")
    master_df = merge_sensor(master_df, labels_df, ['timestamp', 'behavior'], "label")
    
    master_df = master_df.ffill().bfill()
    master_df['label_behavior'] = master_df['label_behavior'].fillna(0).astype(int)
    
    master_df.to_csv(cache_path, index=False)
    return master_df

# ================= 3. PHГӮN ДҗOбә N CHUб»’I (SEGMENTATION) =================
def segment_cow_data(df, window_size, overlap=0.0):
    features = df.drop(columns=['timestamp', 'label_behavior']).values
    labels = df['label_behavior'].values
    step_size = int(window_size * (1 - overlap))
    if step_size < 1:
        step_size = 1
        
    X_list, y_list, w_list = [], [], []
    for i in range(0, len(df) - window_size + 1, step_size):
        X_seg = features[i : i + window_size]
        y_seg = labels[i : i + window_size]
        
        y_mapped = []
        w_seg = []
        for val in y_seg:
            if val == 0:
                y_mapped.append(0)
                w_seg.append(0.0)
            else:
                y_mapped.append(val - 1)
                w_seg.append(1.0)
                
        X_list.append(X_seg)
        y_list.append(y_mapped)
        w_list.append(w_seg)
        
    if len(X_list) == 0:
        return (np.empty((0, window_size, features.shape[1])), 
                np.empty((0, window_size)), 
                np.empty((0, window_size)))
    return np.array(X_list), np.array(y_list), np.array(w_list)

def load_and_segment_split(cow_ids, window_size):
    X_all, y_all, w_all = [], [], []
    for cid in cow_ids:
        df = load_and_align_cow_data(cid)
        X_cow, y_cow, w_cow = segment_cow_data(df, window_size, overlap=0.0)
        if len(X_cow) > 0:
            X_all.append(X_cow)
            y_all.append(y_cow)
            w_all.append(w_cow)
            
    if len(X_all) == 0:
        return (np.empty((0, window_size, N_FEATURES)), 
                np.empty((0, window_size)), 
                np.empty((0, window_size)))
    return np.concatenate(X_all, axis=0), np.concatenate(y_all, axis=0), np.concatenate(w_all, axis=0)

# ================= 4. BILSTM MODEL =================
def build_bilstm_model(seq_len, n_features, n_classes):
    inputs = Input(shape=(seq_len, n_features), name="sensor_input")
    # BiLSTM sб»ӯ dб»Ҙng 128 units -> Дҗбә§u ra 256 chiб»Ғu trб»ұc tiбәҝp trЖ°б»ӣc phГўn loбәЎi
    lstm_out = Bidirectional(LSTM(128, return_sequences=True), name="BiLSTM_Core")(inputs)
    lstm_out_normalized = BatchNormalization()(lstm_out)
    lstm_out_dropout = Dropout(0.3)(lstm_out_normalized)
    outputs = TimeDistributed(Dense(n_classes, activation='softmax'), name="Classifier_Output")(lstm_out_dropout)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# ================= 5. LUб»’NG Xб»¬ LГқ CHГҚNH =================
if __name__ == "__main__":
    print("--- KHб»һI Tбә O CACHE Dб»® LIб»ҶU ---")
    for cid in range(1, 11):
        load_and_align_cow_data(cid)

    # 1. Load vГ  phГўn Д‘oбәЎn dб»Ҝ liб»Үu thГҙ cho Fold 2 (Overlap = 0.0, KhГҙng dГ№ng Augmentation)
    print("\n--- BЖҜб»ҡC 1: LOAD VГҖ PHГӮN ДҗOбә N Dб»® LIб»ҶU FOLD 2 (OVERLAP = 0.0) ---")
    X_train_raw, y_train, w_train = load_and_segment_split(TRAIN_COWS, WINDOW_SIZE)
    X_val_raw, y_val, w_val = load_and_segment_split(VAL_COWS, WINDOW_SIZE)
    X_test_raw, y_test, w_test = load_and_segment_split(TEST_COWS, WINDOW_SIZE)

    print(f"- Train raw shape: {X_train_raw.shape}")
    print(f"- Val raw shape:   {X_val_raw.shape}")
    print(f"- Test raw shape:  {X_test_raw.shape}")

    # 2. Chuбә©n hГіa dб»Ҝ liб»Үu (Fit scaler trГӘn tбәӯp Train cб»§a Fold 2)
    print("\n--- BЖҜб»ҡC 2: CHUбәЁN HГ“A Dб»® LIб»ҶU (FIT TRГҠN TRAIN, TRANSFORM CHO VAL/TEST) ---")
    scaler = StandardScaler()
    scaler.fit(X_train_raw.reshape(-1, N_FEATURES))

    X_train_scaled = scaler.transform(X_train_raw.reshape(-1, N_FEATURES)).reshape(X_train_raw.shape)
    X_val_scaled = scaler.transform(X_val_raw.reshape(-1, N_FEATURES)).reshape(X_val_raw.shape)
    X_test_scaled = scaler.transform(X_test_raw.reshape(-1, N_FEATURES)).reshape(X_test_raw.shape)
    print("- ДҗГЈ chuбә©n hГіa xong 3 tбәӯp dб»Ҝ liб»Үu!")

    # 3. Kiб»ғm tra model checkpoint Д‘ГЈ tб»“n tбәЎi chЖ°a, nбәҝu chЖ°a tiбәҝn hГ nh train nhanh Fold 2
    print("\n--- BЖҜб»ҡC 3: KIб»ӮM TRA MODEL CHECKPOINT ---")
    if not os.path.exists(MODEL_PATH):
        print(f"- KhГҙng tГ¬m thбәҘy {MODEL_PATH}. Tiбәҝn hГ nh huбәҘn luyб»Үn nhanh Fold 2 Д‘б»ғ lбәҘy model tб»‘t nhбәҘt...")
        
        y_train_cat = to_categorical(y_train, num_classes=N_CLASSES)
        y_val_cat = to_categorical(y_val, num_classes=N_CLASSES)
        
        model = build_bilstm_model(WINDOW_SIZE, N_FEATURES, N_CLASSES)
        model.compile(
            loss="categorical_crossentropy",
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
            metrics=["accuracy"]
        )
        
        early_stopping = EarlyStopping(
            monitor="val_loss", 
            patience=8, 
            restore_best_weights=True,
            verbose=1
        )
        
        # LЖ°u checkpoint
        checkpoint = ModelCheckpoint(
            MODEL_PATH, 
            monitor="val_loss", 
            save_best_only=True, 
            verbose=1
        )
        
        model.fit(
            X_train_scaled, y_train_cat,
            sample_weight=w_train,
            epochs=MAX_EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_val_scaled, y_val_cat, w_val),
            callbacks=[early_stopping, checkpoint],
            verbose=1
        )
        print(f"- [SUCCESS] HuбәҘn luyб»Үn xong vГ  Д‘ГЈ lЖ°u model checkpoint tбәЎi: {MODEL_PATH}")
    else:
        print(f"- ДҗГЈ tГ¬m thбәҘy file model checkpoint tбәЎi: {MODEL_PATH}")

    # 4. Khб»ҹi tбәЎo mГҙ hГ¬nh trГӯch xuбәҘt Д‘бә·c trЖ°ng (Feature Extractor)
    print("\n--- BЖҜб»ҡC 4: TбәўI MODEL VГҖ KHб»һI Tбә O FEATURE EXTRACTOR ---")
    full_model = load_model(MODEL_PATH)
    
    # TбәЎo mГҙ hГ¬nh con lбәҘy Д‘бә§u ra tб»« layer "BiLSTM_Core" (256 chiб»Ғu)
    extractor = Model(
        inputs=full_model.input,
        outputs=full_model.get_layer("BiLSTM_Core").output
    )
    print("- Khб»ҹi tбәЎo mГҙ hГ¬nh trГӯch xuбәҘt thГ nh cГҙng!")

    # 5. Thб»ұc hiб»Үn trГӯch xuбәҘt vГ  xuбәҘt Д‘бә·c trЖ°ng ra 3 file .npy
    print("\n--- BЖҜб»ҡC 5: TIбәҫN HГҖNH TRГҚCH XUбәӨT Дҗбә¶C TRЖҜNG BiLSTM 256 CHIб»ҖU ---")
    print("1/3. Дҗang trГӯch xuбәҘt Д‘бә·c trЖ°ng tбәӯp Train...")
    features_train = extractor.predict(X_train_scaled)
    
    print("2/3. Дҗang trГӯch xuбәҘt Д‘бә·c trЖ°ng tбәӯp Val...")
    features_val = extractor.predict(X_val_scaled)
    
    print("3/3. Дҗang trГӯch xuбәҘt Д‘бә·c trЖ°ng tбәӯp Test...")
    features_test = extractor.predict(X_test_scaled)

    # LЖ°u Д‘бә·c trЖ°ng
    print("\n--- BЖҜб»ҡC 6: XUбәӨT Дҗбә¶C TRЖҜNG RA FILE Cб»ҰA FOLD 2 ---")
    np.save("sensor_features_fold2_train.npy", features_train)
    np.save("sensor_features_fold2_val.npy", features_val)
    np.save("sensor_features_fold2_test.npy", features_test)

    print("[SUCCESS] ДҗГЈ xuбәҘt thГ nh cГҙng 3 file Д‘бә·c trЖ°ng:")
    print(f"  + Train: sensor_features_fold2_train.npy | Shape: {features_train.shape}")
    print(f"  + Val:   sensor_features_fold2_val.npy   | Shape: {features_val.shape}")
    print(f"  + Test:  sensor_features_fold2_test.npy  | Shape: {features_test.shape}")

    # LЖ°u nhГЈn vГ  trб»Қng sб»‘ Д‘i kГЁm
    np.save("sensor_labels_fold2_train.npy", y_train)
    np.save("sensor_labels_fold2_val.npy", y_val)
    np.save("sensor_labels_fold2_test.npy", y_test)

    np.save("sensor_weights_fold2_train.npy", w_train)
    np.save("sensor_weights_fold2_val.npy", w_val)
    np.save("sensor_weights_fold2_test.npy", w_test)

    print("\n[SUCCESS] ДҗГЈ lЖ°u nhГЈn vГ  trб»Қng sб»‘ tЖ°ЖЎng б»©ng cho 3 tбәӯp Д‘б»ғ tiб»Үn dГ№ng cho Fusion:")
    print(f"  + Train labels/weights: sensor_labels_fold2_train.npy / sensor_weights_fold2_train.npy")
    print(f"  + Val labels/weights:   sensor_labels_fold2_val.npy / sensor_weights_fold2_val.npy")
    print(f"  + Test labels/weights:  sensor_labels_fold2_test.npy / sensor_weights_fold2_test.npy")

    # In thГҙng tin demo
    print("\n- Дҗбә·c trЖ°ng cб»§a mбә«u Д‘бә§u tiГӘn tбәЎi timestep Д‘бә§u tiГӘn (10 chiб»Ғu Д‘бә§u):")
    print(features_test[0, 0, :10])
```
